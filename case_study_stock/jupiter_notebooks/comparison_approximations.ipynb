```python
import math
import numpy as np

# ------------------------------
# Cubic interpolation functions
# ------------------------------
def lagrange_cubic(pvals, ovals):
    V = np.vstack([pvals**3, pvals**2, pvals, np.ones_like(pvals)]).T
    coeffs = np.linalg.solve(V, ovals)
    return coeffs

def integrate_cubic(a, b, c, d, L, R):
    F = lambda x: (a/4)*x**4 + (b/3)*x**3 + (c/2)*x**2 + d*x
    return F(R) - F(L)

def cubic_octile_method(octiles):
    O = octiles
    p_left  = np.array([1/8, 2/8, 3/8, 4/8])
    p_right = np.array([4/8, 5/8, 6/8, 7/8])
    o_left  = np.array([O[1], O[2], O[3], O[4]])
    o_right = np.array([O[4], O[5], O[6], O[7]])
    a1, b1, c1, d1 = lagrange_cubic(p_left, o_left)
    a2, b2, c2, d2 = lagrange_cubic(p_right, o_right)
    I1 = integrate_cubic(a1, b1, c1, d1, 0.0, 0.5)
    I2 = integrate_cubic(a2, b2, c2, d2, 0.5, 1.0)
    mu = I1 + I2
    h = I2 - I1
    return mu, h

def one_median(octiles):
    result = octiles[4]
    return result, 0

def compute_octiles_uniform():
    res = [0]
    for i in range(1,8):
        p = i/8
        res.append(p)
    return res

def compute_octiles_exponential(my_lambda=1):
    res = [0]
    for i in range(1,8):
        p = i/8
        res.append((-math.log(1-p))/my_lambda)
    return res    

def compute_octiles_laplace(my_mu=0, my_lambda=1):
    res = [0]
    O_1 = my_mu + my_lambda*math.log(2*1/8)
    O_2 = my_mu + my_lambda*math.log(2*2/8)
    O_3 = my_mu + my_lambda*math.log(2*3/8) 
    O_4 = my_mu + my_lambda*math.log(2*4/8)
    O_5 = my_mu - my_lambda*math.log(2 - 2*5/8)
    O_6 = my_mu - my_lambda*math.log(2 - 2*6/8)
    O_7 = my_mu - my_lambda*math.log(2 - 2*7/8)     
    res =[0, O_1, O_2, O_3, O_4, O_5, O_6, O_7]
    return res    

def compute_octiles_pareto(my_alpha = 1.5, my_beta = 1):
    res = [0]
    for i in range(1,8):
        p = i/8
        res.append(my_beta *((1-p)**(-1/my_alpha)))
    return res    

def compare_pareto(my_alpha=1.5, my_beta=1):
    octiles = compute_octiles_pareto(my_alpha, my_beta)
    print("pareto octiles", octiles)
    my_h = my_beta * my_alpha *((2**(1/my_alpha))-1)/(my_alpha -1)
    my_mu = my_alpha * my_beta /(my_alpha -1)
    print("---------------------------")
    print("pareto h:                 ", my_h)
    print("pareto mu:                ", my_mu)
    print("pareto median:            ", one_median(octiles))
    # Cubic approximation
    mu_cubic, h_cubic = cubic_octile_method(octiles)
    print("pareto cubic h:           ", h_cubic)
    print("pareto cubic mu:          ", mu_cubic)
    # Error percentages
    if abs(my_h) > 1e-10:
        h_error = abs((my_h - h_cubic) / my_h) * 100
        print("pareto h error%:          ", h_error)
    if abs(my_mu) > 1e-10:
        mu_error = abs((my_mu - mu_cubic) / my_mu) * 100
        print("pareto mu error%:         ", mu_error)
    print("-------------------------------------------")
    return
 
def compare_laplace(my_mu=0, my_lambda=1):
    octiles = compute_octiles_laplace(my_mu, my_lambda)
    print("laplace octiles", octiles)
    h = my_lambda/math.sqrt(2)
    mu = 0
    dist = "laplace "
    print("---------------------------")
    print(dist + "h:                 ", h)
    print(dist + "mu:                ", mu)
    print(dist + "median:            ", one_median(octiles))
    # Cubic approximation
    mu_cubic, h_cubic = cubic_octile_method(octiles)
    print(dist + "cubic h:           ", h_cubic)
    print(dist + "cubic mu:          ", mu_cubic)
    # Error percentages
    if abs(h) > 1e-10:
        h_error = abs((h - h_cubic) / h) * 100
        print(dist + "h error%:          ", h_error)
    if abs(mu) > 1e-10:
        mu_error = abs((mu - mu_cubic) / mu) * 100
        print(dist + "mu error%:         ", mu_error)
    else:
        print(dist + "mu error%:         0")
    print("----------------------------")
    return   
    
def compare_exponential(my_lambda=1):
    octiles = compute_octiles_exponential(my_lambda)
    print("exponential octiles", octiles)
    h = math.log(2)/my_lambda
    mu = 1/my_lambda
    print("---------------------------")
    print("exp h:                 ", h)
    print("exp mu:                ", mu)
    print("exp median:            ", one_median(octiles))
    # Cubic approximation
    mu_cubic, h_cubic = cubic_octile_method(octiles)
    print("exp cubic h:           ", h_cubic)
    print("exp cubic mu:          ", mu_cubic)
    # Error percentages
    if abs(h) > 1e-10:
        h_error = abs((h - h_cubic) / h) * 100
        print("exp h error%:           ", h_error)
    if abs(mu) > 1e-10:
        mu_error = abs((mu - mu_cubic) / mu) * 100
        print("exp mu error%:          ", mu_error)
    print("----------------------------")
    return

def compare_uniform():
    octiles = compute_octiles_uniform()
    print("uniform octiles", octiles)
    mu, h = 1/2, 1/4
    print("---------------------------")
    print("uniform h:                 ", h)
    print("uniform mu:                ", mu)
    print("uniform median:            ", one_median(octiles))
    # Cubic approximation
    mu_cubic, h_cubic = cubic_octile_method(octiles)
    print("uniform cubic h:           ", h_cubic)
    print("uniform cubic mu:          ", mu_cubic)
    # Error percentages
    if abs(h) > 1e-10:
        h_error = abs((h - h_cubic) / h) * 100
        print("uniform h error%:         ", h_error)
    if abs(mu) > 1e-10:
        mu_error = abs((mu - mu_cubic) / mu) * 100
        print("uniform mu error%:        ", mu_error)
    print("----------------------------")
    return
    
# Run comparisons with similar parameters
compare_laplace(0, 1) 
compare_exponential(1)
compare_uniform()
compare_pareto(200.0, 1/2)
```

```python
import math
import numpy as np
from scipy import stats
from scipy.integrate import quad

# ------------------------------
# Cubic interpolation functions (reused from Cell 1)
# ------------------------------
def lagrange_cubic(pvals, ovals):
    V = np.vstack([pvals**3, pvals**2, pvals, np.ones_like(pvals)]).T
    coeffs = np.linalg.solve(V, ovals)
    return coeffs

def integrate_cubic(a, b, c, d, L, R):
    F = lambda x: (a/4)*x**4 + (b/3)*x**3 + (c/2)*x**2 + d*x
    return F(R) - F(L)

def cubic_octile_method(octiles):
    O = octiles
    # Include Q(0) = O[0] for better boundary handling, especially for distributions where Q(0)=0
    # Left half: use points at 0, 2/8, 3/8, 4/8 for better spacing
    # Right half: use points at 4/8, 5/8, 6/8, 7/8
    p_left  = np.array([0, 2/8, 3/8, 4/8])
    p_right = np.array([4/8, 5/8, 6/8, 7/8])
    o_left  = np.array([O[0], O[2], O[3], O[4]])
    o_right = np.array([O[4], O[5], O[6], O[7]])
    a1, b1, c1, d1 = lagrange_cubic(p_left, o_left)
    a2, b2, c2, d2 = lagrange_cubic(p_right, o_right)
    I1 = integrate_cubic(a1, b1, c1, d1, 0.0, 0.5)
    I2 = integrate_cubic(a2, b2, c2, d2, 0.5, 1.0)
    mu = I1 + I2
    h = I2 - I1
    # Ensure mu is non-negative for non-negative distributions
    # (This is a safeguard, but the interpolation should handle it correctly)
    return mu, h

def one_median(octiles):
    result = octiles[4]
    return result, 0

# ------------------------------
# OCTILE CALCULATIONS for new distributions
# ------------------------------

def compute_octiles_power(alpha=2.0):
    """
    Power distribution on [0,1]: PDF f(x) = alpha * x^(alpha-1), CDF F(x) = x^alpha
    Quantile function: Q(p) = p^(1/alpha)
    """
    res = [0]
    for i in range(1, 8):
        p = i/8
        res.append(p**(1/alpha))
    return res

def compute_octiles_lognormal(mu=0, sigma=1):
    """
    Log-normal distribution: If X ~ LN(mu, sigma^2), then Q(p) = exp(mu + sigma * Phi^(-1)(p))
    where Phi^(-1) is the inverse standard normal CDF
    """
    res = [0]
    for i in range(1, 8):
        p = i/8
        z = stats.norm.ppf(p, loc=0, scale=1)
        res.append(np.exp(mu + sigma * z))
    return res

def compute_octiles_normal(mu=0, sigma=1):
    """
    Normal distribution: Q(p) = mu + sigma * Phi^(-1)(p)
    For p=0 and p=1, we use very small/large finite values instead of infinity
    """
    res = [0]  # Q(0) - use a very negative finite value
    # Use a very small p value for Q(0) approximation
    z0 = stats.norm.ppf(1e-10, loc=0, scale=1)
    res[0] = mu + sigma * z0
    
    for i in range(1, 8):
        p = i/8
        z = stats.norm.ppf(p, loc=0, scale=1)
        res.append(mu + sigma * z)
    
    # Q(1) - use a very large finite value
    z1 = stats.norm.ppf(1 - 1e-10, loc=0, scale=1)
    res.append(mu + sigma * z1)
    return res

def compute_octiles_weibull(shape=2.0, scale=1.0):
    """
    Weibull distribution: Q(p) = scale * (-log(1-p))^(1/shape)
    """
    res = [0]
    for i in range(1, 8):
        p = i/8
        res.append(scale * ((-math.log(1-p))**(1/shape)))
    res.append(np.inf)  # Q(1) is infinite
    return res

# ------------------------------
# TRUE VALUES CALCULATIONS
# ------------------------------

def compute_true_mu_h_power(alpha):
    """True mu and h for power distribution"""
    mu_true = alpha / (alpha + 1)
    # h = ∫₀.₅¹ p^(1/alpha) dp - ∫₀⁰.⁵ p^(1/alpha) dp
    def Q_power(p):
        return p**(1/alpha)
    I1, _ = quad(Q_power, 0.0, 0.5)
    I2, _ = quad(Q_power, 0.5, 1.0)
    h_true = I2 - I1
    return mu_true, h_true

def compute_true_mu_h_lognormal(mu, sigma):
    """True mu and h for log-normal distribution"""
    mu_true = np.exp(mu + sigma**2/2)
    # h = ∫₀.₅¹ Q(p) dp - ∫₀⁰.⁵ Q(p) dp
    def Q_lognormal(p):
        z = stats.norm.ppf(p, loc=0, scale=1)
        return np.exp(mu + sigma * z)
    I1, _ = quad(Q_lognormal, 0.0, 0.5)
    I2, _ = quad(Q_lognormal, 0.5, 1.0)
    h_true = I2 - I1
    return mu_true, h_true

def compute_true_mu_h_normal(mu, sigma):
    """True mu and h for normal distribution"""
    mu_true = mu
    # h = ∫₀.₅¹ Q(p) dp - ∫₀⁰.⁵ Q(p) dp
    def Q_normal(p):
        z = stats.norm.ppf(p, loc=0, scale=1)
        return mu + sigma * z
    I1, _ = quad(Q_normal, 0.0, 0.5)
    I2, _ = quad(Q_normal, 0.5, 1.0)
    h_true = I2 - I1
    return mu_true, h_true

def compute_true_mu_h_weibull(shape, scale):
    """True mu and h for Weibull distribution"""
    from scipy.special import gamma
    mu_true = scale * gamma(1 + 1/shape)
    # h = ∫₀.₅¹ Q(p) dp - ∫₀⁰.⁵ Q(p) dp
    def Q_weibull(p):
        if p >= 1:
            return np.inf
        return scale * ((-math.log(1-p))**(1/shape))
    I1, _ = quad(Q_weibull, 0.0, 0.5)
    I2, _ = quad(Q_weibull, 0.5, 1.0 - 1e-10)  # Avoid p=1
    h_true = I2 - I1
    return mu_true, h_true

# ------------------------------
# COMPARISON FUNCTIONS
# ------------------------------

def compare_power(alpha=2.0):
    octiles = compute_octiles_power(alpha)
    mu_true, h_true = compute_true_mu_h_power(alpha)
    print("power octiles", octiles)
    print("---------------------------")
    print("power h:                 ", h_true)
    print("power mu:                ", mu_true)
    print("power median:            ", one_median(octiles))
    # Cubic approximation
    mu_cubic, h_cubic = cubic_octile_method(octiles)
    print("power cubic h:           ", h_cubic)
    print("power cubic mu:          ", mu_cubic)
    # Error percentages
    if abs(h_true) > 1e-10:
        h_error = abs((h_true - h_cubic) / h_true) * 100
        print("power h error%:          ", h_error)
    if abs(mu_true) > 1e-10:
        mu_error = abs((mu_true - mu_cubic) / mu_true) * 100
        print("power mu error%:         ", mu_error)
    print("-------------------------------------------")
    return

def compare_lognormal(mu=0, sigma=1):
    octiles = compute_octiles_lognormal(mu, sigma)
    mu_true, h_true = compute_true_mu_h_lognormal(mu, sigma)
    print("lognormal octiles", octiles)
    print("---------------------------")
    print("lognormal h:                 ", h_true)
    print("lognormal mu:                ", mu_true)
    print("lognormal median:            ", one_median(octiles))
    # Cubic approximation
    mu_cubic, h_cubic = cubic_octile_method(octiles)
    print("lognormal cubic h:           ", h_cubic)
    print("lognormal cubic mu:          ", mu_cubic)
    # Error percentages
    if abs(h_true) > 1e-10:
        h_error = abs((h_true - h_cubic) / h_true) * 100
        print("lognormal h error%:          ", h_error)
    if abs(mu_true) > 1e-10:
        mu_error = abs((mu_true - mu_cubic) / mu_true) * 100
        print("lognormal mu error%:         ", mu_error)
    print("-------------------------------------------")
    return

def compare_normal(mu=0, sigma=1):
    octiles = compute_octiles_normal(mu, sigma)
    mu_true, h_true = compute_true_mu_h_normal(mu, sigma)
    print("normal octiles", octiles)
    print("---------------------------")
    print("normal h:                 ", h_true)
    print("normal mu:                ", mu_true)
    print("normal median:            ", one_median(octiles))
    # Cubic approximation
    mu_cubic, h_cubic = cubic_octile_method(octiles)
    print("normal cubic h:           ", h_cubic)
    print("normal cubic mu:          ", abs(mu_cubic))
    # Error percentages
    if abs(h_true) > 1e-10:
        h_error = abs((h_true - h_cubic) / h_true) * 100
        print("normal h error%:          ", h_error)
    if abs(mu_true) > 1e-10:
        mu_error = abs((mu_true - abs(mu_cubic)) / mu_true) * 100
        print("normal mu error%:         ", mu_error)
    print("-------------------------------------------")
    return

def compare_weibull(shape=2.0, scale=1.0):
    octiles = compute_octiles_weibull(shape, scale)
    mu_true, h_true = compute_true_mu_h_weibull(shape, scale)
    print("weibull octiles", octiles)
    print("---------------------------")
    print("weibull h:                 ", h_true)
    print("weibull mu:                ", mu_true)
    print("weibull median:            ", one_median(octiles))
    # Cubic approximation (only use finite octiles)
    octiles_finite = [q for q in octiles if np.isfinite(q)]
    if len(octiles_finite) >= 8:
        mu_cubic, h_cubic = cubic_octile_method(octiles_finite)
        print("weibull cubic h:           ", h_cubic)
        print("weibull cubic mu:          ", mu_cubic)
        # Error percentages
        if abs(h_true) > 1e-10:
            h_error = abs((h_true - h_cubic) / h_true) * 100
            print("weibull h error%:          ", h_error)
        if abs(mu_true) > 1e-10:
            mu_error = abs((mu_true - mu_cubic) / mu_true) * 100
            print("weibull mu error%:         ", mu_error)
    else:
        print("weibull cubic: Cannot compute (too many infinite values)")
    print("-------------------------------------------")
    return

# ------------------------------
# RUN ALL COMPARISONS
# ------------------------------

print("=" * 60)
print("POWER DISTRIBUTION (alpha=2.0)")
print("=" * 60)
compare_power(alpha=2.0)

print("\n" + "=" * 60)
print("LOG-NORMAL DISTRIBUTION (mu=0, sigma=1)")
print("=" * 60)
compare_lognormal(mu=0, sigma=1)

print("\n" + "=" * 60)
print("NORMAL DISTRIBUTION (mu=0, sigma=1)")
print("=" * 60)
compare_normal(mu=0, sigma=1)

print("\n" + "=" * 60)
print("WEIBULL DISTRIBUTION (shape=2.0, scale=1.0)")
print("=" * 60)
compare_weibull(shape=2.0, scale=1.0)
```

```python
import math
import numpy as np
from scipy import stats
from scipy.integrate import quad

# ------------------------------
# Quadratic interpolation functions (reused from previous cell)
# ------------------------------
def lagrange_quadratic(pvals, ovals):
    # Fit quadratic polynomial ax^2 + bx + c through 3 points
    V = np.vstack([pvals**2, pvals, np.ones_like(pvals)]).T
    coeffs = np.linalg.solve(V, ovals)
    return coeffs

def integrate_quadratic(a, b, c, L, R):
    F = lambda x: (a/3)*x**3 + (b/2)*x**2 + c*x
    return F(R) - F(L)

def quadratic_octile_method(octiles):
    O = octiles
    # Use 3 points for quadratic interpolation on each half
    # Left half: use points at 2/8, 3/8, 4/8
    p_left  = np.array([2/8, 3/8, 4/8])
    o_left  = np.array([O[2], O[3], O[4]])
    # Right half: use points at 4/8, 5/8, 7/8
    p_right = np.array([4/8, 5/8, 7/8])
    o_right = np.array([O[4], O[5], O[7]])
    a1, b1, c1 = lagrange_quadratic(p_left, o_left)
    a2, b2, c2 = lagrange_quadratic(p_right, o_right)
    I1 = integrate_quadratic(a1, b1, c1, 0.0, 0.5)
    I2 = integrate_quadratic(a2, b2, c2, 0.5, 1.0)
    mu = I1 + I2
    h = I2 - I1
    return mu, h

def one_median(octiles):
    result = octiles[4]
    return result, 0

# ------------------------------
# OCTILE CALCULATIONS for new distributions
# ------------------------------

def compute_octiles_power(alpha=2.0):
    """
    Power distribution on [0,1]: PDF f(x) = alpha * x^(alpha-1), CDF F(x) = x^alpha
    Quantile function: Q(p) = p^(1/alpha)
    """
    res = [0]
    for i in range(1, 8):
        p = i/8
        res.append(p**(1/alpha))
    return res

def compute_octiles_lognormal(mu=0, sigma=1):
    """
    Log-normal distribution: If X ~ LN(mu, sigma^2), then Q(p) = exp(mu + sigma * Phi^(-1)(p))
    where Phi^(-1) is the inverse standard normal CDF
    """
    res = [0]
    for i in range(1, 8):
        p = i/8
        z = stats.norm.ppf(p, loc=0, scale=1)
        res.append(np.exp(mu + sigma * z))
    return res

def compute_octiles_normal(mu=0, sigma=1):
    """
    Normal distribution: Q(p) = mu + sigma * Phi^(-1)(p)
    For p=0 and p=1, we use very small/large finite values instead of infinity
    """
    res = [0]  # Q(0) - use a very negative finite value
    # Use a very small p value for Q(0) approximation
    z0 = stats.norm.ppf(1e-10, loc=0, scale=1)
    res[0] = mu + sigma * z0
    
    for i in range(1, 8):
        p = i/8
        z = stats.norm.ppf(p, loc=0, scale=1)
        res.append(mu + sigma * z)
    
    # Q(1) - use a very large finite value
    z1 = stats.norm.ppf(1 - 1e-10, loc=0, scale=1)
    res.append(mu + sigma * z1)
    return res

def compute_octiles_weibull(shape=2.0, scale=1.0):
    """
    Weibull distribution: Q(p) = scale * (-log(1-p))^(1/shape)
    """
    res = [0]
    for i in range(1, 8):
        p = i/8
        res.append(scale * ((-math.log(1-p))**(1/shape)))
    res.append(np.inf)  # Q(1) is infinite
    return res

# ------------------------------
# TRUE VALUES CALCULATIONS
# ------------------------------

def compute_true_mu_h_power(alpha):
    """True mu and h for power distribution"""
    mu_true = alpha / (alpha + 1)
    # h = ∫₀.₅¹ p^(1/alpha) dp - ∫₀⁰.⁵ p^(1/alpha) dp
    def Q_power(p):
        return p**(1/alpha)
    I1, _ = quad(Q_power, 0.0, 0.5)
    I2, _ = quad(Q_power, 0.5, 1.0)
    h_true = I2 - I1
    return mu_true, h_true

def compute_true_mu_h_lognormal(mu, sigma):
    """True mu and h for log-normal distribution"""
    mu_true = np.exp(mu + sigma**2/2)
    # h = ∫₀.₅¹ Q(p) dp - ∫₀⁰.⁵ Q(p) dp
    def Q_lognormal(p):
        z = stats.norm.ppf(p, loc=0, scale=1)
        return np.exp(mu + sigma * z)
    I1, _ = quad(Q_lognormal, 0.0, 0.5)
    I2, _ = quad(Q_lognormal, 0.5, 1.0)
    h_true = I2 - I1
    return mu_true, h_true

def compute_true_mu_h_normal(mu, sigma):
    """True mu and h for normal distribution"""
    mu_true = mu
    # h = ∫₀.₅¹ Q(p) dp - ∫₀⁰.⁵ Q(p) dp
    def Q_normal(p):
        z = stats.norm.ppf(p, loc=0, scale=1)
        return mu + sigma * z
    I1, _ = quad(Q_normal, 0.0, 0.5)
    I2, _ = quad(Q_normal, 0.5, 1.0)
    h_true = I2 - I1
    return mu_true, h_true

def compute_true_mu_h_weibull(shape, scale):
    """True mu and h for Weibull distribution"""
    from scipy.special import gamma
    mu_true = scale * gamma(1 + 1/shape)
    # h = ∫₀.₅¹ Q(p) dp - ∫₀⁰.⁵ Q(p) dp
    def Q_weibull(p):
        if p >= 1:
            return np.inf
        return scale * ((-math.log(1-p))**(1/shape))
    I1, _ = quad(Q_weibull, 0.0, 0.5)
    I2, _ = quad(Q_weibull, 0.5, 1.0 - 1e-10)  # Avoid p=1
    h_true = I2 - I1
    return mu_true, h_true

# ------------------------------
# COMPARISON FUNCTIONS
# ------------------------------

def compare_power(alpha=2.0):
    octiles = compute_octiles_power(alpha)
    mu_true, h_true = compute_true_mu_h_power(alpha)
    print("power octiles", octiles)
    print("---------------------------")
    print("power h:                 ", h_true)
    print("power mu:                ", mu_true)
    print("power median:            ", one_median(octiles))
    # Quadratic approximation
    mu_quadratic, h_quadratic = quadratic_octile_method(octiles)
    print("power quadratic h:           ", h_quadratic)
    print("power quadratic mu:          ", mu_quadratic)
    # Error percentages
    if abs(h_true) > 1e-10:
        h_error = abs((h_true - h_quadratic) / h_true) * 100
        print("power h error%:          ", h_error)
    if abs(mu_true) > 1e-10:
        mu_error = abs((mu_true - mu_quadratic) / mu_true) * 100
        print("power mu error%:         ", mu_error)
    print("-------------------------------------------")
    return

def compare_lognormal(mu=0, sigma=1):
    octiles = compute_octiles_lognormal(mu, sigma)
    mu_true, h_true = compute_true_mu_h_lognormal(mu, sigma)
    print("lognormal octiles", octiles)
    print("---------------------------")
    print("lognormal h:                 ", h_true)
    print("lognormal mu:                ", mu_true)
    print("lognormal median:            ", one_median(octiles))
    # Quadratic approximation
    mu_quadratic, h_quadratic = quadratic_octile_method(octiles)
    print("lognormal quadratic h:           ", h_quadratic)
    print("lognormal quadratic mu:          ", mu_quadratic)
    # Error percentages
    if abs(h_true) > 1e-10:
        h_error = abs((h_true - h_quadratic) / h_true) * 100
        print("lognormal h error%:          ", h_error)
    if abs(mu_true) > 1e-10:
        mu_error = abs((mu_true - mu_quadratic) / mu_true) * 100
        print("lognormal mu error%:         ", mu_error)
    print("-------------------------------------------")
    return

def compare_normal(mu=0, sigma=1):
    octiles = compute_octiles_normal(mu, sigma)
    mu_true, h_true = compute_true_mu_h_normal(mu, sigma)
    print("normal octiles", octiles)
    print("---------------------------")
    print("normal h:                 ", h_true)
    print("normal mu:                ", mu_true)
    print("normal median:            ", one_median(octiles))
    # Quadratic approximation
    mu_quadratic, h_quadratic = quadratic_octile_method(octiles)
    print("normal quadratic h:           ", h_quadratic)
    print("normal quadratic mu:          ", mu_quadratic)
    # Error percentages
    if abs(h_true) > 1e-10:
        h_error = abs((h_true - h_quadratic) / h_true) * 100
        print("normal h error%:          ", h_error)
        if abs(mu_true) > 1e-10:
            mu_error = abs((mu_true - mu_quadratic) / mu_true) * 100
            print("normal mu error%:         ", mu_error)
    else:
        # When mu_true is 0, print absolute error instead
            mu_abs_error = abs(mu_true - mu_quadratic)
            print("normal mu absolute error:  ", mu_abs_error)
            print("normal mu error%:         N/A (mu_true = 0)")
    print("-------------------------------------------")
    return

def compare_weibull(shape=2.0, scale=1.0):
    octiles = compute_octiles_weibull(shape, scale)
    mu_true, h_true = compute_true_mu_h_weibull(shape, scale)
    print("weibull octiles", octiles)
    print("---------------------------")
    print("weibull h:                 ", h_true)
    print("weibull mu:                ", mu_true)
    print("weibull median:            ", one_median(octiles))
    # Quadratic approximation (only use finite octiles)
    octiles_finite = [q for q in octiles if np.isfinite(q)]
    if len(octiles_finite) >= 8:
        mu_quadratic, h_quadratic = quadratic_octile_method(octiles_finite)
        print("weibull quadratic h:           ", h_quadratic)
        print("weibull quadratic mu:          ", mu_quadratic)
        # Error percentages
        if abs(h_true) > 1e-10:
            h_error = abs((h_true - h_quadratic) / h_true) * 100
            print("weibull h error%:          ", h_error)
        if abs(mu_true) > 1e-10:
            mu_error = abs((mu_true - mu_quadratic) / mu_true) * 100
            print("weibull mu error%:         ", mu_error)
    else:
        print("weibull quadratic: Cannot compute (too many infinite values)")
    print("-------------------------------------------")
    return

# ------------------------------
# RUN ALL COMPARISONS
# ------------------------------

print("=" * 60)
print("POWER DISTRIBUTION (alpha=2.0)")
print("=" * 60)
compare_power(alpha=2.0)

print("\n" + "=" * 60)
print("LOG-NORMAL DISTRIBUTION (mu=0, sigma=1)")
print("=" * 60)
compare_lognormal(mu=0, sigma=1)

print("\n" + "=" * 60)
print("NORMAL DISTRIBUTION (mu=0, sigma=1)")
print("=" * 60)
compare_normal(mu=0, sigma=1)

print("\n" + "=" * 60)
print("WEIBULL DISTRIBUTION (shape=2.0, scale=1.0)")
print("=" * 60)
compare_weibull(shape=2.0, scale=1.0)
```

```python
import math
import numpy as np

# ------------------------------
# Quadratic endpoint estimation with trapezoid approximation
# ------------------------------
def quadratic_endpoint_trapezoid_method(octiles):
    """
    Uses quadratic interpolation to estimate endpoints, then trapezoid approximation.
    C_L(0) = 3O_1 - 3O_2 + O_3
    C_R(1) = O_5 - 3O_6 + 3O_7
    I1 = (5O_1 - O_2 + 3O_3 + O_4) / 16
    I2 = (O_4 + 3O_5 - O_6 + 5O_7) / 16
    mu = I1 + I2
    h = I2 - I1
    """
    O = octiles
    O_1, O_2, O_3, O_4 = O[1], O[2], O[3], O[4]
    O_5, O_6, O_7 = O[5], O[6], O[7]
    
    # Estimate endpoints using quadratic interpolation
    C_L_0 = 3*O_1 - 3*O_2 + O_3
    C_R_1 = O_5 - 3*O_6 + 3*O_7
    
    # Trapezoid approximation with estimated endpoints
    I1 = (5*O_1 - O_2 + 3*O_3 + O_4) / 16.0
    I2 = (O_4 + 3*O_5 - O_6 + 5*O_7) / 16.0
    
    mu = I1 + I2
    h = I2 - I1
    
    return mu, h

def one_median(octiles):
    result = octiles[4]
    return result, 0

def compute_octiles_uniform():
    res = [0]
    for i in range(1,8):
        p = i/8
        res.append(p)
    return res

def compute_octiles_exponential(my_lambda=1):
    res = [0]
    for i in range(1,8):
        p = i/8
        res.append((-math.log(1-p))/my_lambda)
    return res    

def compute_octiles_laplace(my_mu=0, my_lambda=1):
    res = [0]
    O_1 = my_mu + my_lambda*math.log(2*1/8)
    O_2 = my_mu + my_lambda*math.log(2*2/8)
    O_3 = my_mu + my_lambda*math.log(2*3/8) 
    O_4 = my_mu + my_lambda*math.log(2*4/8)
    O_5 = my_mu - my_lambda*math.log(2 - 2*5/8)
    O_6 = my_mu - my_lambda*math.log(2 - 2*6/8)
    O_7 = my_mu - my_lambda*math.log(2 - 2*7/8)     
    res =[0, O_1, O_2, O_3, O_4, O_5, O_6, O_7]
    return res    

def compute_octiles_pareto(my_alpha = 1.5, my_beta = 1):
    res = [0]
    for i in range(1,8):
        p = i/8
        res.append(my_beta *((1-p)**(-1/my_alpha)))
    return res    

def compare_pareto(my_alpha=1.5, my_beta=1):
    octiles = compute_octiles_pareto(my_alpha, my_beta)
    print("pareto octiles", octiles)
    my_h = my_beta * my_alpha *((2**(1/my_alpha))-1)/(my_alpha -1)
    my_mu = my_alpha * my_beta /(my_alpha -1)
    print("---------------------------")
    print("pareto h:                 ", my_h)
    print("pareto mu:                ", my_mu)
    print("pareto median:            ", one_median(octiles))
    # Quadratic endpoint trapezoid approximation
    mu_quad_trap, h_quad_trap = quadratic_endpoint_trapezoid_method(octiles)
    print("pareto quadratic endpoint trapezoid h:           ", h_quad_trap)
    print("pareto quadratic endpoint trapezoid mu:          ", mu_quad_trap)
    # Error percentages
    if abs(my_h) > 1e-10:
        h_error = abs((my_h - h_quad_trap) / my_h) * 100
        print("pareto h error%:          ", h_error)
    if abs(my_mu) > 1e-10:
        mu_error = abs((my_mu - mu_quad_trap) / my_mu) * 100
        print("pareto mu error%:         ", mu_error)
    print("-------------------------------------------")
    return
 
def compare_laplace(my_mu=0, my_lambda=1):
    octiles = compute_octiles_laplace(my_mu, my_lambda)
    print("laplace octiles", octiles)
    h = my_lambda/math.sqrt(2)
    mu = 0
    dist = "laplace "
    print("---------------------------")
    print(dist + "h:                 ", h)
    print(dist + "mu:                ", mu)
    print(dist + "median:            ", one_median(octiles))
    # Quadratic endpoint trapezoid approximation
    mu_quad_trap, h_quad_trap = quadratic_endpoint_trapezoid_method(octiles)
    print(dist + "quadratic endpoint trapezoid h:           ", h_quad_trap)
    print(dist + "quadratic endpoint trapezoid mu:          ", mu_quad_trap)
    # Error percentages
    if abs(h) > 1e-10:
        h_error = abs((h - h_quad_trap) / h) * 100
        print(dist + "h error%:          ", h_error)
    if abs(mu) > 1e-10:
        mu_error = abs((mu - mu_quad_trap) / mu) * 100
        print(dist + "mu error%:         ", mu_error)
    else:
        print(dist + "mu error%:         0")
    print("----------------------------")
    return   
    
def compare_exponential(my_lambda=1):
    octiles = compute_octiles_exponential(my_lambda)
    print("exponential octiles", octiles)
    h = math.log(2)/my_lambda
    mu = 1/my_lambda
    print("---------------------------")
    print("exp h:                 ", h)
    print("exp mu:                ", mu)
    print("exp median:            ", one_median(octiles))
    # Quadratic endpoint trapezoid approximation
    mu_quad_trap, h_quad_trap = quadratic_endpoint_trapezoid_method(octiles)
    print("exp quadratic endpoint trapezoid h:           ", h_quad_trap)
    print("exp quadratic endpoint trapezoid mu:          ", mu_quad_trap)
    # Error percentages
    if abs(h) > 1e-10:
        h_error = abs((h - h_quad_trap) / h) * 100
        print("exp h error%:           ", h_error)
    if abs(mu) > 1e-10:
        mu_error = abs((mu - mu_quad_trap) / mu) * 100
        print("exp mu error%:          ", mu_error)
    print("----------------------------")
    return

def compare_uniform():
    octiles = compute_octiles_uniform()
    print("uniform octiles", octiles)
    mu, h = 1/2, 1/4
    print("---------------------------")
    print("uniform h:                 ", h)
    print("uniform mu:                ", mu)
    print("uniform median:            ", one_median(octiles))
    # Quadratic endpoint trapezoid approximation
    mu_quad_trap, h_quad_trap = quadratic_endpoint_trapezoid_method(octiles)
    print("uniform quadratic endpoint trapezoid h:           ", h_quad_trap)
    print("uniform quadratic endpoint trapezoid mu:          ", mu_quad_trap)
    # Error percentages
    if abs(h) > 1e-10:
        h_error = abs((h - h_quad_trap) / h) * 100
        print("uniform h error%:         ", h_error)
    if abs(mu) > 1e-10:
        mu_error = abs((mu - mu_quad_trap) / mu) * 100
        print("uniform mu error%:        ", mu_error)
    print("----------------------------")
    return
    
# Run comparisons with similar parameters
compare_laplace(0, 1) 
compare_exponential(1)
compare_uniform()
compare_pareto(2.0, 1/2)
```
