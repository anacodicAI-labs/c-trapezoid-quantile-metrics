```python
import math

def compute_error(method, h_exact, h_app, mu_exact, mu_app):
    error_1 = 100*(h_exact-h_app)/h_exact
    error_2 = 100*(mu_exact-mu_app)/mu_exact
    suffix = ""
    print(method, " h_ex: ", h_exact, " h_app: ", h_app, 
          " rel_error:", round(error_1, 2))
    print(method, " mu_ex: ", mu_exact, " mu_app: ", mu_app, 
          " rel_error:", round(error_2, 2))
    return

def compute_I_error(method, I_1_exact, I_1_app, I_2_exact, I_2_app):
    error_1 = 100*(I_1_exact - I_1_app)/h_exact
    error_2 = 100*(I_2_exact - I_2_app)/mu_exact
    print(method, " I_1: ", I_1_exact, " I_1_app: ", I_1_app, 
          " rel_error:", round(error_1, 2))
    print(method, " I_2: ", I_2_exact, " I_2_app: ", I_2_app, 
          " rel_error:", round(error_2, 2), "\n")
    return

my_lambda = 1
O_1 = -math.log(7/8)/my_lambda; O_2 = -math.log(6/8)/my_lambda; O_3 = -math.log(5/8)/my_lambda; 
O_4 = -math.log(4/8)/my_lambda; O_5 = -math.log(3/8)/my_lambda; O_6 = -math.log(2/8)/my_lambda; 
O_7 = -math.log(1/8)/my_lambda
O_8 = -O_4 + 4*O_5 -6*O_6 + 4*O_7
O_0 = 4*O_1 -6*O_2 +4*O_3 -O_4

I_1_exact = (1-math.log(2))/(2*my_lambda); I_2_exact = (1+math.log(2))/(2*my_lambda)
h_exact = math.log(2)/my_lambda; mu_exact = 1/my_lambda

I_1 = O_4/2; I_2 = O_4/2;
compute_error("Midpoint: ", h_exact, I_2 - I_1, mu_exact, I_1 + I_2)
compute_I_error("Midpoint_I: ", I_1_exact, I_1, I_2_exact, I_2)

I_1 = O_2/2; I_2 = O_6/2;
compute_error("Two Quartiles: ", h_exact, I_2 - I_1, mu_exact, I_1 + I_2)
compute_I_error("Two Quartiles: ", I_1_exact, I_1, I_2_exact, I_2)

I_1 = (O_2+O_4)/4; I_2 = (O_4 + O_6)/4;
compute_error("Three Quartiles: ", h_exact, I_2-I_1, mu_exact, I_1 + I_2)
compute_I_error("Three Quartiles: ", I_1_exact, I_1, I_2_exact, I_2)


I_1 = (O_1 + O_3)/4; I_2 = (O_5 + O_7)/4;
compute_error("Four Octiles: ", h_exact, I_2 - I_1, mu_exact, I_1 + I_2)
compute_I_error("Four Octiles: ", I_1_exact, I_1, I_2_exact, I_2)

I_1 =(O_1 + O_2 + O_3 + O_4)/8; I_2 = (O_4+ O_5 +O_6 + O_7)/8;
compute_error("All Octiles: ", h_exact, I_2 - I_1, mu_exact, I_1 + I_2)
compute_I_error("All Octiles: ", I_1_exact, I_1, I_2_exact, I_2)


I_1 =(3*O_1 +2*O_2 +2*O_3 + O_4)/16; I_2 = (O_4 +2*O_5 +2*O_6 +3* O_7)/16;
mu_app = I_1 + I_2;  h_app = I_2 - I_1;
compute_error("rectange plus trapezoid : ", h_exact, I_2 - I_1, mu_exact, I_1 + I_2)
compute_I_error("rectangle plus trapezoid: ", I_1_exact, I_1, I_2_exact, I_2)


a = 11*math.log(2)+6*math.log(3)-4*math.log(5)-4 * math.log(7)
b = 11*math.log(2) - 4*math.log(3)
print("a: ", a, "   b: ", b, math.e**(-3.23))

I_1 = (3*O_1 -2*O_2 +3*O_3)/8;    I_2= (3*O_5 -2*O_6 + 3*O_7)/8;
compute_error("cubic-trapezoid : ", h_exact, I_2 - I_1, mu_exact, I_1 + I_2)
compute_I_error("cubic -trapezoid: ", I_1_exact, I_1, I_2_exact, I_2)


I_1 = (O_0+O_2)/8 + (O_2+O_4)/8;
I_2 = (O_4 + O_6)/8 +(O_6 + O_8)/8;
compute_error("cubic-Q-trapezoid : ", h_exact, I_2 - I_1, mu_exact, I_1 + I_2)
compute_I_error("cubic -Q-trapezoid: ", I_1_exact, I_1, I_2_exact, I_2)




# cubic + simpson
I_1 = (O_0 + 4*O_2 + O_4)/12; I_2 = (O_4 + 4*O_6 + O_8)/12;
compute_error("cubic-simpson : ", h_exact, I_2 - I_1, mu_exact, I_1 + I_2)
compute_I_error("cubic plus simpson: ", I_1_exact, I_1, I_2_exact, I_2)


# cubic + 2-simpson
I_1 = ((O_0 + 4*O_1 + O_2)+ (O_2 + 4*O_3 + O_4))/24;
I_2 =((O_4 + 4*O_5 + O_6) +(O_6 + 4*O_7 + O_8))/24;
compute_error("cubic-2-simpson : ", h_exact, I_2 - I_1, mu_exact, I_1 + I_2)
compute_I_error("cubic plus 2-simpson: ", I_1_exact, I_1, I_2_exact, I_2)
```
