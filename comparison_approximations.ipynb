```python
import math
import numpy as np
from statistics import NormalDist

a = 3* math.log(3)/(2*math.log(2))-2
print(a)
b=math.log(2)/(5*math.log(2)-3*math.log(3))
print(b)


beta = 1
alpha = 1.5

def trapezoid(octiles):
    O_1, O_2, O_3 = octiles[1], octiles[2], octiles[3]
    O_4, O_5, O_6, O_7 = octiles[4], octiles[5], octiles[6], octiles[7]
    first_half = O_1/8 + (O_1 + O_2)/16 + (O_2 + O_3)/16 + (O_3 + O_4)/16 
    second_half = (O_4+O_5)/16 +(O_5+O_6)/16 + (O_6 + O_7)/16 + O_7/8
    result = first_half + second_half
    return result

def simpson(octiles):
    O_1, O_2, O_3 = octiles[1], octiles[2], octiles[3]
    O_4, O_5, O_6, O_7 = octiles[4], octiles[5], octiles[6], octiles[7]
    first_half = O_1/4 + (O_2 + 4*O_3 + O_4)/24
    second_half = (O_4 + 4*O_5 + O_6)/24 + O_7/4
    result = first_half + second_half
    return result

def simpson_3(octiles):
    O_1, O_2, O_3 = octiles[1], octiles[2], octiles[3]
    O_4, O_5, O_6, O_7 = octiles[4], octiles[5], octiles[6], octiles[7]
    left_part = (O_1 + 4*O_2 + O_3)/24
    middle_part = (O_3 + 4*O_4 + O_5)/24
    right_part = (O_5 + 4*O_6 + O_7)/24
    result = O_1/8 + left_part + middle_part + right_part + O_7/8
    return result

def simpson_composite(octiles):
    O_1, O_2, O_3 = octiles[1], octiles[2], octiles[3]
    O_4, O_5, O_6, O_7 = octiles[4], octiles[5], octiles[6], octiles[7]
    first_half = O_1/8 + (O_1 + 3*O_2 + 3*O_3 + O_4)*3/64
    second_half = (O_4 + 3*O_5 + 3*O_6 + O_7)*3/64 + O_7/8
    result = first_half + second_half
    return result

def one_median(octiles):
    result = octiles[4]
    return result, 0

def two_quartiles(octiles):
    O_2 = octiles[2]
    O_6 = octiles[6]
    mu_two = (O_2 + O_6)/2
    h_two = (O_6 - O_2)/2
    return mu_two, h_two

def four_quartiles(octiles):
    O_2 = octiles[2]
    O_4 = octiles[4]
    O_6 = octiles[6]
    mu_four = (O_2 + 2*O_4 + O_4)/4
    h_four  = (O_6 - O_2)/2    
    return mu_four, h_four

def four_octiles(octiles):
    O_1 = octiles[1]
    O_3 = octiles[3]
    O_5 = octiles[5]
    O_7 = octiles[7]
    mu_four = (O_1 + O_3 + O_5 + O_7)/4
    h_four  = ((O_5 + O_7)-(O_1+O_3))/4   
    return mu_four, h_four

def all_octiles(octiles):
    O_1, O_2, O_3 = octiles[1], octiles[2], octiles[3]
    O_4, O_5, O_6, O_7 = octiles[4], octiles[5], octiles[6], octiles[7]
    first_half  = (O_1 + O_2 + O_3 + O_4)/8
    second_half = (O_4 + O_5 + O_6 + O_7)/8
    h = second_half - first_half
    result = first_half + second_half, h
    
    return result

def two_rect_two_trap(octiles):
    O_1, O_2, O_3 = octiles[1], octiles[2], octiles[3]
    O_4, O_5, O_6, O_7 = octiles[4], octiles[5], octiles[6], octiles[7]
    first_half  = 5*O_1/16 +3*O_4/16
    second_half = 3*O_4/16 + 5*O_7/16
    result = first_half + second_half
    return result

def two_rect_one_trap(octiles):
    O_1, O_2, O_3 = octiles[1], octiles[2], octiles[3]
    O_4, O_5, O_6, O_7 = octiles[4], octiles[5], octiles[6], octiles[7]
    result = (O_1 + O_7)/2 
    return result

def one_rect_three_trapezoids(octiles):
    O_1, O_2, O_3 = octiles[1], octiles[2], octiles[3]
    O_4, O_5, O_6, O_7 = octiles[4], octiles[5], octiles[6], octiles[7]
    I_1 = (2*O_1 + 2*O_2 +2*O_3 + O_4)/16.0
    I_2 =(O_4 + 2*O_5 + 2*O_6 + 3*O_7)/16.0
    new_mu = I_1 + I_2
    new_h  = I_2 - I_1
    return new_mu, new_h


def one_triangle_three_trapezoids(octiles):
    O_1, O_2, O_3 = octiles[1], octiles[2], octiles[3]
    O_4, O_5, O_6, O_7 = octiles[4], octiles[5], octiles[6], octiles[7]
    I_1 = (3*O_1 + 2*O_2 +2*O_3 + O_4)/16.0
    I_2 =(O_4 + 2*O_5 + O_6 + 4*O_7)/16.0
    new_mu = I_1 + I_2
    new_h  = I_2 - I_1
    return new_mu, new_h






def compute_octiles_uniform():
    res = [0]
    for i in range(1,8):
        p = i/8
        res.append(p)
    return res

def compute_octiles_exponential(my_lambda=1):
    res = [0]
    for i in range(1,8):
        p = i/8
        res.append((-math.log(1-p))/my_lambda)
    return res    

def compute_octiles_laplace(my_mu=0, my_lambda=1):
    res = [0]
    O_1 = my_mu + my_lambda*math.log(2*1/8)
    O_2 = my_mu + my_lambda*math.log(2*2/8)
    O_3 = my_mu + my_lambda*math.log(2*3/8) 
    O_4 = my_mu + my_lambda*math.log(2*4/8)
    O_5 = my_mu - my_lambda*math.log(2 - 2*5/8)
    O_6 = my_mu - my_lambda*math.log(2 - 2*6/8)
    O_7 = my_mu - my_lambda*math.log(2 - 2*7/8)     
    res =[0, O_1, O_2, O_3, O_4, O_5, O_6, O_7]
    return res    

def compute_octiles_pareto(my_alpha = 1.5, my_beta = 1):
    res = [0]
    for i in range(1,8):
        p = i/8
        res.append(my_beta *((1-p)**(-1/my_alpha)))
    return res    

STD_NORMAL = NormalDist()

def integrate_quantile(q_func, start, end, num=5000):
    ps = np.linspace(start, end, num)
    values = np.array([q_func(p) for p in ps])
    return np.trapz(values, ps)

def compute_true_h(q_func, eps=1e-6, num=5000):
    lower = integrate_quantile(q_func, eps, 0.5, num)
    upper = integrate_quantile(q_func, 0.5, 1 - eps, num)
    return upper - lower

def compute_octiles_power(alpha=2.0):
    res = [0]
    for i in range(1,8):
        p = i/8
        res.append(p**(1/alpha))
    return res

def compute_octiles_lognormal(log_mu=0, log_sigma=1):
    res = [0]
    for i in range(1,8):
        p = i/8
        z = STD_NORMAL.inv_cdf(p)
        res.append(math.exp(log_mu + log_sigma * z))
    return res

def compute_octiles_normal(my_mu=0, my_sigma=1):
    res = [0]
    normal_dist = NormalDist(my_mu, my_sigma)
    for i in range(1,8):
        p = i/8
        res.append(normal_dist.inv_cdf(p))
    return res

def compute_octiles_weibull(shape=2.0, scale=1.0):
    res = [0]
    for i in range(1,8):
        p = i/8
        res.append(scale * ((-math.log(1-p))**(1/shape)))
    return res


def compare_pareto(my_alpha=1.5, my_beta=1):
    octiles = compute_octiles_pareto(my_alpha, my_beta)
    print("pareto octiles", octiles)
    my_h = my_beta * my_alpha *((2**(1/my_alpha))-1)/(my_alpha -1)
    my_mu = my_alpha * my_beta /(my_alpha -1)
    print("---------------------------")
    print("pareto h:                 ", my_h)
    print("pareto mu:                ", my_mu)
    print("pareto median:            ", one_median(octiles))
    
    # Calculate all methods
    mu_2q, h_2q = two_quartiles(octiles)
    mu_4q, h_4q = four_quartiles(octiles)
    mu_4o, h_4o = four_octiles(octiles)
    mu_ao, h_ao = all_octiles(octiles)
    mu_trap = trapezoid(octiles)
    mu_2r2t = two_rect_two_trap(octiles)
    mu_2r1t = two_rect_one_trap(octiles)
    mu_simp = simpson(octiles)
    mu_simp3 = simpson_3(octiles)
    mu_simpc = simpson_composite(octiles)
    mu_1r3t, h_1r3t = one_rect_three_trapezoids(octiles)
    mu_1t3t, h_1t3t = one_triangle_three_trapezoids(octiles)
    
    # Print values
    print("pareto 2-quartiles:       ", two_quartiles(octiles))
    print("pareto 4-quartiles        ", four_quartiles(octiles))
    print("pareto 4-octiles :        ", four_octiles(octiles))
    print("pareto all-octiles :      ", all_octiles(octiles))
    print("pareto trap mu:           ", trapezoid(octiles))
    print("pareto 2-rec-2-trap       ", two_rect_two_trap(octiles))
    print("pareto 2-rec-1-trap       ", two_rect_one_trap(octiles))
    print("pareto simpson:           ", simpson(octiles))
    print("pareto simpson_3:         ", simpson_3(octiles))
    print("pareto simpson_composite: ", simpson_composite(octiles))
    
    # Print 1 rec 3 trap with errors
    print("pareto 1 rec 3 trap mu: ", mu_1r3t)
    print("pareto 1 rec 3 trap h: ", h_1r3t)
    if abs(my_mu) > 1e-10:
        print("pareto 1 rec 3 trap mu error%:", abs((my_mu - mu_1r3t) / my_mu) * 100)
    if abs(my_h) > 1e-10:
        print("pareto 1 rec 3 trap h error%: ", abs((my_h - h_1r3t) / my_h) * 100)
    
    print("pareto 1 tri 3 trap mu: ", one_triangle_three_trapezoids(octiles)[0])
    print("pareto 1 tri 3 trap h: ", one_triangle_three_trapezoids(octiles)[1])
    print("-------------------------------------------")
    return
 
def compare_laplace(my_mu=0, my_lambda=1):
    octiles = compute_octiles_laplace(my_mu, my_lambda)
    print("laplace octiles", octiles)
    h = my_lambda/math.sqrt(2)
    mu = 0
    dist = "laplace "
    print("---------------------------")
    print(dist + "h:                 ", h)
    print(dist + "mu:                ", mu)
    print(dist + "median:            ", one_median(octiles))
    
    # Calculate all methods
    mu_2q, h_2q = two_quartiles(octiles)
    mu_4q, h_4q = four_quartiles(octiles)
    mu_4o, h_4o = four_octiles(octiles)
    mu_ao, h_ao = all_octiles(octiles)
    mu_trap = trapezoid(octiles)
    mu_2r2t = two_rect_two_trap(octiles)
    mu_2r1t = two_rect_one_trap(octiles)
    mu_simp = simpson(octiles)
    mu_simp3 = simpson_3(octiles)
    mu_simpc = simpson_composite(octiles)
    mu_1r3t, h_1r3t = one_rect_three_trapezoids(octiles)
    mu_1t3t, h_1t3t = one_triangle_three_trapezoids(octiles)
    
    # Print values
    print(dist + "2-quartiles:       ", two_quartiles(octiles))
    print(dist + "4-quartiles        ", four_quartiles(octiles))
    print(dist + "4-octiles :        ", four_octiles(octiles))
    print(dist + "all-octiles :      ", all_octiles(octiles))
    print(dist + "trap mu:           ", trapezoid(octiles))
    print(dist + "2-rec-2-trap       ", two_rect_two_trap(octiles))
    print(dist + "2-rec-1-trap       ", two_rect_one_trap(octiles))
    print(dist + "simpson:           ", simpson(octiles))
    print(dist + "simpson_3:         ", simpson_3(octiles))
    print(dist + "simpson_composite: ", simpson_composite(octiles))
    
    # Print 1 rec 3 trap with errors
    print(dist + "1 rec 3 trap mu: ", mu_1r3t)
    print(dist + "1 rec 3 trap h: ", h_1r3t)
    if abs(mu) > 1e-10:
        print(dist + "1 rec 3 trap mu error%:", abs((mu - mu_1r3t) / mu) * 100)
    else:
        print(dist + "1 rec 3 trap mu error%: N/A (actual mu is 0)")
    if abs(h) > 1e-10:
        print(dist + "1 rec 3 trap h error%: ", abs((h - h_1r3t) / h) * 100)
    
    print(dist + "1 tri 3 trap mu: ", one_triangle_three_trapezoids(octiles)[0])
    print(dist + "1 tri 3 trap h: ",  one_triangle_three_trapezoids(octiles)[1])
    print("----------------------------")
    return   
    
    
    
def compare_exponential(my_lambda=1):
    octiles = compute_octiles_exponential(my_lambda)
    print("exponential octiles", octiles)
    h = math.log(2)/my_lambda
    mu = 1/my_lambda
    dist ="exp "
    print("---------------------------")
    print("exp h:                 ", h)
    print("exp mu:                ", mu)
    print("exp median:            ", one_median(octiles))
    
    # Calculate all methods
    mu_2q, h_2q = two_quartiles(octiles)
    mu_4q, h_4q = four_quartiles(octiles)
    mu_4o, h_4o = four_octiles(octiles)
    mu_ao, h_ao = all_octiles(octiles)
    mu_trap = trapezoid(octiles)
    mu_2r2t = two_rect_two_trap(octiles)
    mu_2r1t = two_rect_one_trap(octiles)
    mu_simp = simpson(octiles)
    mu_simp3 = simpson_3(octiles)
    mu_simpc = simpson_composite(octiles)
    mu_1r3t, h_1r3t = one_rect_three_trapezoids(octiles)
    mu_1t3t, h_1t3t = one_triangle_three_trapezoids(octiles)
    
    # Print values
    print("exp 2-quartiles:       ", two_quartiles(octiles))
    print("exp 4-quartiles        ", four_quartiles(octiles))
    print("exp 4-octiles :        ", four_octiles(octiles))
    print("exp all-octiles :      ", all_octiles(octiles))
    print("exp trap mu:           ", trapezoid(octiles))
    print("exp 2-rec-2-trap       ", two_rect_two_trap(octiles))
    print("exp 2-rec-1-trap       ", two_rect_one_trap(octiles))
    print("exp simpson:           ", simpson(octiles))
    print("exp simpson_3:         ", simpson_3(octiles))
    print("exp simpson_composite: ", simpson_composite(octiles))
    
    # Print 1 rec 3 trap with errors
    print("exp 1 rec 3 trap mu: ", mu_1r3t)
    print("exp 1 rec 3 trap h: ", h_1r3t)
    if abs(mu) > 1e-10:
        print("exp 1 rec 3 trap mu error%:", abs((mu - mu_1r3t) / mu) * 100)
    if abs(h) > 1e-10:
        print("exp 1 rec 3 trap h error%: ", abs((h - h_1r3t) / h) * 100)
    
    print(dist + " 1 tri 3 trap mu: ", one_triangle_three_trapezoids(octiles)[0])
    print(dist + " 1 tri 3 trap h: ",  one_triangle_three_trapezoids(octiles)[1])
    print("----------------------------")
    return




def compare_uniform():
    octiles = compute_octiles_uniform()
    print("uniform octiles", octiles)
    mu, h = 1/2, 1/4
    print("---------------------------")
    print("uniform h:                 ", h)
    print("uniform mu:                ", mu)
    print("uniform median:            ", one_median(octiles))
    
    # Calculate all methods
    mu_2q, h_2q = two_quartiles(octiles)
    mu_4q, h_4q = four_quartiles(octiles)
    mu_4o, h_4o = four_octiles(octiles)
    mu_ao, h_ao = all_octiles(octiles)
    mu_trap = trapezoid(octiles)
    mu_2r2t = two_rect_two_trap(octiles)
    mu_2r1t = two_rect_one_trap(octiles)
    mu_simp = simpson(octiles)
    mu_simp3 = simpson_3(octiles)
    mu_simpc = simpson_composite(octiles)
    
    # Print values
    print("uniform 2-quartiles:       ", two_quartiles(octiles))
    print("uniform 4-quartiles        ", four_quartiles(octiles))
    print("uniform 4-octiles :        ", four_octiles(octiles))
    print("uniform all-octiles :      ", all_octiles(octiles))
    print("uniform trap mu:           ", trapezoid(octiles))
    print("uniform 2-rec-2-trap       ", two_rect_two_trap(octiles))
    print("uniform 2-rec-1-trap       ", two_rect_one_trap(octiles))
    print("uniform simpson:           ", simpson(octiles))
    print("uniform simpson_3:         ", simpson_3(octiles))
    print("uniform simpson_composite: ", simpson_composite(octiles))
    
    # Calculate and print 1 rec 3 trap with errors
    mu_1r3t, h_1r3t = one_rect_three_trapezoids(octiles)
    print("uniform 1 rec 3 trap mu: ", mu_1r3t)
    print("uniform 1 rec 3 trap h: ", h_1r3t)
    if abs(mu) > 1e-10:
        print("uniform 1 rec 3 trap mu error%:", abs((mu - mu_1r3t) / mu) * 100)
    if abs(h) > 1e-10:
        print("uniform 1 rec 3 trap h error%: ", abs((h - h_1r3t) / h) * 100)
    print("----------------------------")
    return


def compare_power(alpha=2.0):
    octiles = compute_octiles_power(alpha)
    print("power octiles", octiles)
    mu = alpha / (alpha + 1)
    h = compute_true_h(lambda p, a=alpha: p**(1/a))
    dist = "power "
    print("---------------------------")
    print(dist + "h:                 ", h)
    print(dist + "mu:                ", mu)
    print(dist + "median:            ", one_median(octiles))
    
    mu_2q, h_2q = two_quartiles(octiles)
    mu_4q, h_4q = four_quartiles(octiles)
    mu_4o, h_4o = four_octiles(octiles)
    mu_ao, h_ao = all_octiles(octiles)
    mu_trap = trapezoid(octiles)
    mu_2r2t = two_rect_two_trap(octiles)
    mu_2r1t = two_rect_one_trap(octiles)
    mu_simp = simpson(octiles)
    mu_simp3 = simpson_3(octiles)
    mu_simpc = simpson_composite(octiles)
    mu_1r3t, h_1r3t = one_rect_three_trapezoids(octiles)
    mu_1t3t, h_1t3t = one_triangle_three_trapezoids(octiles)
    
    print(dist + "2-quartiles:       ", two_quartiles(octiles))
    print(dist + "4-quartiles        ", four_quartiles(octiles))
    print(dist + "4-octiles :        ", four_octiles(octiles))
    print(dist + "all-octiles :      ", all_octiles(octiles))
    print(dist + "trap mu:           ", trapezoid(octiles))
    print(dist + "2-rec-2-trap       ", two_rect_two_trap(octiles))
    print(dist + "2-rec-1-trap       ", two_rect_one_trap(octiles))
    print(dist + "simpson:           ", simpson(octiles))
    print(dist + "simpson_3:         ", simpson_3(octiles))
    print(dist + "simpson_composite: ", simpson_composite(octiles))
    print(dist + "1 rec 3 trap mu: ", mu_1r3t)
    print(dist + "1 rec 3 trap h: ", h_1r3t)
    if abs(mu) > 1e-10:
        print(dist + "1 rec 3 trap mu error%:", abs((mu - mu_1r3t) / mu) * 100)
    if abs(h) > 1e-10:
        print(dist + "1 rec 3 trap h error%: ", abs((h - h_1r3t) / h) * 100)
    print(dist + "1 tri 3 trap mu: ", mu_1t3t)
    print(dist + "1 tri 3 trap h: ", h_1t3t)
    print("----------------------------")
    return


def compare_lognormal(log_mu=0, log_sigma=1):
    octiles = compute_octiles_lognormal(log_mu, log_sigma)
    print("lognormal octiles", octiles)
    mu = math.exp(log_mu + (log_sigma**2)/2)
    h = compute_true_h(lambda p, m=log_mu, s=log_sigma: math.exp(m + s * STD_NORMAL.inv_cdf(p)))
    dist = "lognormal "
    print("---------------------------")
    print(dist + "h:                 ", h)
    print(dist + "mu:                ", mu)
    print(dist + "median:            ", one_median(octiles))
    
    mu_2q, h_2q = two_quartiles(octiles)
    mu_4q, h_4q = four_quartiles(octiles)
    mu_4o, h_4o = four_octiles(octiles)
    mu_ao, h_ao = all_octiles(octiles)
    mu_trap = trapezoid(octiles)
    mu_2r2t = two_rect_two_trap(octiles)
    mu_2r1t = two_rect_one_trap(octiles)
    mu_simp = simpson(octiles)
    mu_simp3 = simpson_3(octiles)
    mu_simpc = simpson_composite(octiles)
    mu_1r3t, h_1r3t = one_rect_three_trapezoids(octiles)
    mu_1t3t, h_1t3t = one_triangle_three_trapezoids(octiles)
    
    print(dist + "2-quartiles:       ", two_quartiles(octiles))
    print(dist + "4-quartiles        ", four_quartiles(octiles))
    print(dist + "4-octiles :        ", four_octiles(octiles))
    print(dist + "all-octiles :      ", all_octiles(octiles))
    print(dist + "trap mu:           ", trapezoid(octiles))
    print(dist + "2-rec-2-trap       ", two_rect_two_trap(octiles))
    print(dist + "2-rec-1-trap       ", two_rect_one_trap(octiles))
    print(dist + "simpson:           ", simpson(octiles))
    print(dist + "simpson_3:         ", simpson_3(octiles))
    print(dist + "simpson_composite: ", simpson_composite(octiles))
    print(dist + "1 rec 3 trap mu: ", mu_1r3t)
    print(dist + "1 rec 3 trap h: ", h_1r3t)
    if abs(mu) > 1e-10:
        print(dist + "1 rec 3 trap mu error%:", abs((mu - mu_1r3t) / mu) * 100)
    if abs(h) > 1e-10:
        print(dist + "1 rec 3 trap h error%: ", abs((h - h_1r3t) / h) * 100)
    print(dist + "1 tri 3 trap mu: ", mu_1t3t)
    print(dist + "1 tri 3 trap h: ", h_1t3t)
    print("----------------------------")
    return


def compare_normal_dist(my_mu=0, my_sigma=1):
    octiles = compute_octiles_normal(my_mu, my_sigma)
    print("normal octiles", octiles)
    norm = NormalDist(my_mu, my_sigma)
    h = compute_true_h(lambda p, nd=norm: nd.inv_cdf(p))
    mu = my_mu
    dist = "normal "
    print("---------------------------")
    print(dist + "h:                 ", h)
    print(dist + "mu:                ", mu)
    print(dist + "median:            ", one_median(octiles))
    
    mu_2q, h_2q = two_quartiles(octiles)
    mu_4q, h_4q = four_quartiles(octiles)
    mu_4o, h_4o = four_octiles(octiles)
    mu_ao, h_ao = all_octiles(octiles)
    mu_trap = trapezoid(octiles)
    mu_2r2t = two_rect_two_trap(octiles)
    mu_2r1t = two_rect_one_trap(octiles)
    mu_simp = simpson(octiles)
    mu_simp3 = simpson_3(octiles)
    mu_simpc = simpson_composite(octiles)
    mu_1r3t, h_1r3t = one_rect_three_trapezoids(octiles)
    mu_1t3t, h_1t3t = one_triangle_three_trapezoids(octiles)
    
    print(dist + "2-quartiles:       ", two_quartiles(octiles))
    print(dist + "4-quartiles        ", four_quartiles(octiles))
    print(dist + "4-octiles :        ", four_octiles(octiles))
    print(dist + "all-octiles :      ", all_octiles(octiles))
    print(dist + "trap mu:           ", trapezoid(octiles))
    print(dist + "2-rec-2-trap       ", two_rect_two_trap(octiles))
    print(dist + "2-rec-1-trap       ", two_rect_one_trap(octiles))
    print(dist + "simpson:           ", simpson(octiles))
    print(dist + "simpson_3:         ", simpson_3(octiles))
    print(dist + "simpson_composite: ", simpson_composite(octiles))
    print(dist + "1 rec 3 trap mu: ", mu_1r3t)
    print(dist + "1 rec 3 trap h: ", h_1r3t)
    if abs(mu) > 1e-10:
        print(dist + "1 rec 3 trap mu error%:", abs((mu - mu_1r3t) / mu) * 100)
    if abs(h) > 1e-10:
        print(dist + "1 rec 3 trap h error%: ", abs((h - h_1r3t) / h) * 100)
    print(dist + "1 tri 3 trap mu: ", mu_1t3t)
    print(dist + "1 tri 3 trap h: ", h_1t3t)
    print("----------------------------")
    return


def compare_weibull(shape=2.0, scale=1.0):
    octiles = compute_octiles_weibull(shape, scale)
    print("weibull octiles", octiles)
    mu = scale * math.gamma(1 + 1/shape)
    h = compute_true_h(lambda p, s=shape, b=scale: b * ((-math.log(1-p))**(1/s)), eps=1e-6)
    dist = "weibull "
    print("---------------------------")
    print(dist + "h:                 ", h)
    print(dist + "mu:                ", mu)
    print(dist + "median:            ", one_median(octiles))
    
    mu_2q, h_2q = two_quartiles(octiles)
    mu_4q, h_4q = four_quartiles(octiles)
    mu_4o, h_4o = four_octiles(octiles)
    mu_ao, h_ao = all_octiles(octiles)
    mu_trap = trapezoid(octiles)
    mu_2r2t = two_rect_two_trap(octiles)
    mu_2r1t = two_rect_one_trap(octiles)
    mu_simp = simpson(octiles)
    mu_simp3 = simpson_3(octiles)
    mu_simpc = simpson_composite(octiles)
    mu_1r3t, h_1r3t = one_rect_three_trapezoids(octiles)
    mu_1t3t, h_1t3t = one_triangle_three_trapezoids(octiles)
    
    print(dist + "2-quartiles:       ", two_quartiles(octiles))
    print(dist + "4-quartiles        ", four_quartiles(octiles))
    print(dist + "4-octiles :        ", four_octiles(octiles))
    print(dist + "all-octiles :      ", all_octiles(octiles))
    print(dist + "trap mu:           ", trapezoid(octiles))
    print(dist + "2-rec-2-trap       ", two_rect_two_trap(octiles))
    print(dist + "2-rec-1-trap       ", two_rect_one_trap(octiles))
    print(dist + "simpson:           ", simpson(octiles))
    print(dist + "simpson_3:         ", simpson_3(octiles))
    print(dist + "simpson_composite: ", simpson_composite(octiles))
    print(dist + "1 rec 3 trap mu: ", mu_1r3t)
    print(dist + "1 rec 3 trap h: ", h_1r3t)
    if abs(mu) > 1e-10:
        print(dist + "1 rec 3 trap mu error%:", abs((mu - mu_1r3t) / mu) * 100)
    if abs(h) > 1e-10:
        print(dist + "1 rec 3 trap h error%: ", abs((h - h_1r3t) / h) * 100)
    print(dist + "1 tri 3 trap mu: ", mu_1t3t)
    print(dist + "1 tri 3 trap h: ", h_1t3t)
    print("----------------------------")
    return


compare_laplace(0, 1) 
compare_exponential(1)
compare_uniform()
compare_pareto(2.0, 1/2)
compare_power(2.0)
compare_lognormal(0, 1)
compare_normal_dist(0, 1)
compare_weibull(2.0, 1.0)



```

```python
import math
import numpy as np

# ------------------------------
# Cubic interpolation functions
# ------------------------------
def lagrange_cubic(pvals, ovals):
    V = np.vstack([pvals**3, pvals**2, pvals, np.ones_like(pvals)]).T
    coeffs = np.linalg.solve(V, ovals)
    return coeffs

def integrate_cubic(a, b, c, d, L, R):
    F = lambda x: (a/4)*x**4 + (b/3)*x**3 + (c/2)*x**2 + d*x
    return F(R) - F(L)

def cubic_octile_method(octiles):
    O = octiles
    p_left  = np.array([1/8, 2/8, 3/8, 4/8])
    p_right = np.array([4/8, 5/8, 6/8, 7/8])
    o_left  = np.array([O[1], O[2], O[3], O[4]])
    o_right = np.array([O[4], O[5], O[6], O[7]])
    a1, b1, c1, d1 = lagrange_cubic(p_left, o_left)
    a2, b2, c2, d2 = lagrange_cubic(p_right, o_right)
    I1 = integrate_cubic(a1, b1, c1, d1, 0.0, 0.5)
    I2 = integrate_cubic(a2, b2, c2, d2, 0.5, 1.0)
    mu = I1 + I2
    h = I2 - I1
    return mu, h

def one_median(octiles):
    result = octiles[4]
    return result, 0

def compute_octiles_uniform():
    res = [0]
    for i in range(1,8):
        p = i/8
        res.append(p)
    return res

def compute_octiles_exponential(my_lambda=1):
    res = [0]
    for i in range(1,8):
        p = i/8
        res.append((-math.log(1-p))/my_lambda)
    return res    

def compute_octiles_laplace(my_mu=0, my_lambda=1):
    res = [0]
    O_1 = my_mu + my_lambda*math.log(2*1/8)
    O_2 = my_mu + my_lambda*math.log(2*2/8)
    O_3 = my_mu + my_lambda*math.log(2*3/8) 
    O_4 = my_mu + my_lambda*math.log(2*4/8)
    O_5 = my_mu - my_lambda*math.log(2 - 2*5/8)
    O_6 = my_mu - my_lambda*math.log(2 - 2*6/8)
    O_7 = my_mu - my_lambda*math.log(2 - 2*7/8)     
    res =[0, O_1, O_2, O_3, O_4, O_5, O_6, O_7]
    return res    

def compute_octiles_pareto(my_alpha = 1.5, my_beta = 1):
    res = [0]
    for i in range(1,8):
        p = i/8
        res.append(my_beta *((1-p)**(-1/my_alpha)))
    return res    

def compare_pareto(my_alpha=1.5, my_beta=1):
    octiles = compute_octiles_pareto(my_alpha, my_beta)
    print("pareto octiles", octiles)
    my_h = my_beta * my_alpha *((2**(1/my_alpha))-1)/(my_alpha -1)
    my_mu = my_alpha * my_beta /(my_alpha -1)
    print("---------------------------")
    print("pareto h:                 ", my_h)
    print("pareto mu:                ", my_mu)
    print("pareto median:            ", one_median(octiles))
    # Cubic approximation
    mu_cubic, h_cubic = cubic_octile_method(octiles)
    print("pareto cubic h:           ", h_cubic)
    print("pareto cubic mu:          ", mu_cubic)
    # Error percentages
    if abs(my_h) > 1e-10:
        h_error = abs((my_h - h_cubic) / my_h) * 100
        print("pareto h error%:          ", h_error)
    if abs(my_mu) > 1e-10:
        mu_error = abs((my_mu - mu_cubic) / my_mu) * 100
        print("pareto mu error%:         ", mu_error)
    print("-------------------------------------------")
    return
 
def compare_laplace(my_mu=0, my_lambda=1):
    octiles = compute_octiles_laplace(my_mu, my_lambda)
    print("laplace octiles", octiles)
    h = my_lambda/math.sqrt(2)
    mu = 0
    dist = "laplace "
    print("---------------------------")
    print(dist + "h:                 ", h)
    print(dist + "mu:                ", mu)
    print(dist + "median:            ", one_median(octiles))
    # Cubic approximation
    mu_cubic, h_cubic = cubic_octile_method(octiles)
    print(dist + "cubic h:           ", h_cubic)
    print(dist + "cubic mu:          ", mu_cubic)
    # Error percentages
    if abs(h) > 1e-10:
        h_error = abs((h - h_cubic) / h) * 100
        print(dist + "h error%:          ", h_error)
    if abs(mu) > 1e-10:
        mu_error = abs((mu - mu_cubic) / mu) * 100
        print(dist + "mu error%:         ", mu_error)
    else:
        print(dist + "mu error%:         0")
    print("----------------------------")
    return   
    
def compare_exponential(my_lambda=1):
    octiles = compute_octiles_exponential(my_lambda)
    print("exponential octiles", octiles)
    h = math.log(2)/my_lambda
    mu = 1/my_lambda
    print("---------------------------")
    print("exp h:                 ", h)
    print("exp mu:                ", mu)
    print("exp median:            ", one_median(octiles))
    # Cubic approximation
    mu_cubic, h_cubic = cubic_octile_method(octiles)
    print("exp cubic h:           ", h_cubic)
    print("exp cubic mu:          ", mu_cubic)
    # Error percentages
    if abs(h) > 1e-10:
        h_error = abs((h - h_cubic) / h) * 100
        print("exp h error%:           ", h_error)
    if abs(mu) > 1e-10:
        mu_error = abs((mu - mu_cubic) / mu) * 100
        print("exp mu error%:          ", mu_error)
    print("----------------------------")
    return

def compare_uniform():
    octiles = compute_octiles_uniform()
    print("uniform octiles", octiles)
    mu, h = 1/2, 1/4
    print("---------------------------")
    print("uniform h:                 ", h)
    print("uniform mu:                ", mu)
    print("uniform median:            ", one_median(octiles))
    # Cubic approximation
    mu_cubic, h_cubic = cubic_octile_method(octiles)
    print("uniform cubic h:           ", h_cubic)
    print("uniform cubic mu:          ", mu_cubic)
    # Error percentages
    if abs(h) > 1e-10:
        h_error = abs((h - h_cubic) / h) * 100
        print("uniform h error%:         ", h_error)
    if abs(mu) > 1e-10:
        mu_error = abs((mu - mu_cubic) / mu) * 100
        print("uniform mu error%:        ", mu_error)
    print("----------------------------")
    return
    
# Run comparisons with similar parameters
compare_laplace(0, 1) 
compare_exponential(1)
compare_uniform()
compare_pareto(2.0, 1/2)
```

```python
import math
import numpy as np
from scipy import stats
from scipy.integrate import quad

# ------------------------------
# Cubic interpolation functions (reused from Cell 1)
# ------------------------------
def lagrange_cubic(pvals, ovals):
    V = np.vstack([pvals**3, pvals**2, pvals, np.ones_like(pvals)]).T
    coeffs = np.linalg.solve(V, ovals)
    return coeffs

def integrate_cubic(a, b, c, d, L, R):
    F = lambda x: (a/4)*x**4 + (b/3)*x**3 + (c/2)*x**2 + d*x
    return F(R) - F(L)

def cubic_octile_method(octiles):
    O = octiles
    # Include Q(0) = O[0] for better boundary handling, especially for distributions where Q(0)=0
    # Left half: use points at 0, 2/8, 3/8, 4/8 for better spacing
    # Right half: use points at 4/8, 5/8, 6/8, 7/8
    p_left  = np.array([0, 2/8, 3/8, 4/8])
    p_right = np.array([4/8, 5/8, 6/8, 7/8])
    o_left  = np.array([O[0], O[2], O[3], O[4]])
    o_right = np.array([O[4], O[5], O[6], O[7]])
    a1, b1, c1, d1 = lagrange_cubic(p_left, o_left)
    a2, b2, c2, d2 = lagrange_cubic(p_right, o_right)
    I1 = integrate_cubic(a1, b1, c1, d1, 0.0, 0.5)
    I2 = integrate_cubic(a2, b2, c2, d2, 0.5, 1.0)
    mu = I1 + I2
    h = I2 - I1
    # Ensure mu is non-negative for non-negative distributions
    # (This is a safeguard, but the interpolation should handle it correctly)
    return mu, h

def one_median(octiles):
    result = octiles[4]
    return result, 0

# ------------------------------
# OCTILE CALCULATIONS for new distributions
# ------------------------------

def compute_octiles_power(alpha=2.0):
    """
    Power distribution on [0,1]: PDF f(x) = alpha * x^(alpha-1), CDF F(x) = x^alpha
    Quantile function: Q(p) = p^(1/alpha)
    """
    res = [0]
    for i in range(1, 8):
        p = i/8
        res.append(p**(1/alpha))
    return res

def compute_octiles_lognormal(mu=0, sigma=1):
    """
    Log-normal distribution: If X ~ LN(mu, sigma^2), then Q(p) = exp(mu + sigma * Phi^(-1)(p))
    where Phi^(-1) is the inverse standard normal CDF
    """
    res = [0]
    for i in range(1, 8):
        p = i/8
        z = stats.norm.ppf(p, loc=0, scale=1)
        res.append(np.exp(mu + sigma * z))
    return res

def compute_octiles_normal(mu=0, sigma=1):
    """
    Normal distribution: Q(p) = mu + sigma * Phi^(-1)(p)
    For p=0 and p=1, we use very small/large finite values instead of infinity
    """
    res = [0]  # Q(0) - use a very negative finite value
    # Use a very small p value for Q(0) approximation
    z0 = stats.norm.ppf(1e-10, loc=0, scale=1)
    res[0] = mu + sigma * z0
    
    for i in range(1, 8):
        p = i/8
        z = stats.norm.ppf(p, loc=0, scale=1)
        res.append(mu + sigma * z)
    
    # Q(1) - use a very large finite value
    z1 = stats.norm.ppf(1 - 1e-10, loc=0, scale=1)
    res.append(mu + sigma * z1)
    return res

def compute_octiles_weibull(shape=2.0, scale=1.0):
    """
    Weibull distribution: Q(p) = scale * (-log(1-p))^(1/shape)
    """
    res = [0]
    for i in range(1, 8):
        p = i/8
        res.append(scale * ((-math.log(1-p))**(1/shape)))
    res.append(np.inf)  # Q(1) is infinite
    return res

# ------------------------------
# TRUE VALUES CALCULATIONS
# ------------------------------

def compute_true_mu_h_power(alpha):
    """True mu and h for power distribution"""
    mu_true = alpha / (alpha + 1)
    # h = ∫₀.₅¹ p^(1/alpha) dp - ∫₀⁰.⁵ p^(1/alpha) dp
    def Q_power(p):
        return p**(1/alpha)
    I1, _ = quad(Q_power, 0.0, 0.5)
    I2, _ = quad(Q_power, 0.5, 1.0)
    h_true = I2 - I1
    return mu_true, h_true

def compute_true_mu_h_lognormal(mu, sigma):
    """True mu and h for log-normal distribution"""
    mu_true = np.exp(mu + sigma**2/2)
    # h = ∫₀.₅¹ Q(p) dp - ∫₀⁰.⁵ Q(p) dp
    def Q_lognormal(p):
        z = stats.norm.ppf(p, loc=0, scale=1)
        return np.exp(mu + sigma * z)
    I1, _ = quad(Q_lognormal, 0.0, 0.5)
    I2, _ = quad(Q_lognormal, 0.5, 1.0)
    h_true = I2 - I1
    return mu_true, h_true

def compute_true_mu_h_normal(mu, sigma):
    """True mu and h for normal distribution"""
    mu_true = mu
    # h = ∫₀.₅¹ Q(p) dp - ∫₀⁰.⁵ Q(p) dp
    def Q_normal(p):
        z = stats.norm.ppf(p, loc=0, scale=1)
        return mu + sigma * z
    I1, _ = quad(Q_normal, 0.0, 0.5)
    I2, _ = quad(Q_normal, 0.5, 1.0)
    h_true = I2 - I1
    return mu_true, h_true

def compute_true_mu_h_weibull(shape, scale):
    """True mu and h for Weibull distribution"""
    from scipy.special import gamma
    mu_true = scale * gamma(1 + 1/shape)
    # h = ∫₀.₅¹ Q(p) dp - ∫₀⁰.⁵ Q(p) dp
    def Q_weibull(p):
        if p >= 1:
            return np.inf
        return scale * ((-math.log(1-p))**(1/shape))
    I1, _ = quad(Q_weibull, 0.0, 0.5)
    I2, _ = quad(Q_weibull, 0.5, 1.0 - 1e-10)  # Avoid p=1
    h_true = I2 - I1
    return mu_true, h_true

# ------------------------------
# COMPARISON FUNCTIONS
# ------------------------------

def compare_power(alpha=2.0):
    octiles = compute_octiles_power(alpha)
    mu_true, h_true = compute_true_mu_h_power(alpha)
    print("power octiles", octiles)
    print("---------------------------")
    print("power h:                 ", h_true)
    print("power mu:                ", mu_true)
    print("power median:            ", one_median(octiles))
    # Cubic approximation
    mu_cubic, h_cubic = cubic_octile_method(octiles)
    print("power cubic h:           ", h_cubic)
    print("power cubic mu:          ", mu_cubic)
    # Error percentages
    if abs(h_true) > 1e-10:
        h_error = abs((h_true - h_cubic) / h_true) * 100
        print("power h error%:          ", h_error)
    if abs(mu_true) > 1e-10:
        mu_error = abs((mu_true - mu_cubic) / mu_true) * 100
        print("power mu error%:         ", mu_error)
    print("-------------------------------------------")
    return

def compare_lognormal(mu=0, sigma=1):
    octiles = compute_octiles_lognormal(mu, sigma)
    mu_true, h_true = compute_true_mu_h_lognormal(mu, sigma)
    print("lognormal octiles", octiles)
    print("---------------------------")
    print("lognormal h:                 ", h_true)
    print("lognormal mu:                ", mu_true)
    print("lognormal median:            ", one_median(octiles))
    # Cubic approximation
    mu_cubic, h_cubic = cubic_octile_method(octiles)
    print("lognormal cubic h:           ", h_cubic)
    print("lognormal cubic mu:          ", mu_cubic)
    # Error percentages
    if abs(h_true) > 1e-10:
        h_error = abs((h_true - h_cubic) / h_true) * 100
        print("lognormal h error%:          ", h_error)
    if abs(mu_true) > 1e-10:
        mu_error = abs((mu_true - mu_cubic) / mu_true) * 100
        print("lognormal mu error%:         ", mu_error)
    print("-------------------------------------------")
    return

def compare_normal(mu=0, sigma=1):
    octiles = compute_octiles_normal(mu, sigma)
    mu_true, h_true = compute_true_mu_h_normal(mu, sigma)
    print("normal octiles", octiles)
    print("---------------------------")
    print("normal h:                 ", h_true)
    print("normal mu:                ", mu_true)
    print("normal median:            ", one_median(octiles))
    # Cubic approximation
    mu_cubic, h_cubic = cubic_octile_method(octiles)
    print("normal cubic h:           ", h_cubic)
    print("normal cubic mu:          ", abs(mu_cubic))
    # Error percentages
    if abs(h_true) > 1e-10:
        h_error = abs((h_true - h_cubic) / h_true) * 100
        print("normal h error%:          ", h_error)
    if abs(mu_true) > 1e-10:
        mu_error = abs((mu_true - abs(mu_cubic)) / mu_true) * 100
        print("normal mu error%:         ", mu_error)
    print("-------------------------------------------")
    return

def compare_weibull(shape=2.0, scale=1.0):
    octiles = compute_octiles_weibull(shape, scale)
    mu_true, h_true = compute_true_mu_h_weibull(shape, scale)
    print("weibull octiles", octiles)
    print("---------------------------")
    print("weibull h:                 ", h_true)
    print("weibull mu:                ", mu_true)
    print("weibull median:            ", one_median(octiles))
    # Cubic approximation (only use finite octiles)
    octiles_finite = [q for q in octiles if np.isfinite(q)]
    if len(octiles_finite) >= 8:
        mu_cubic, h_cubic = cubic_octile_method(octiles_finite)
        print("weibull cubic h:           ", h_cubic)
        print("weibull cubic mu:          ", mu_cubic)
        # Error percentages
        if abs(h_true) > 1e-10:
            h_error = abs((h_true - h_cubic) / h_true) * 100
            print("weibull h error%:          ", h_error)
        if abs(mu_true) > 1e-10:
            mu_error = abs((mu_true - mu_cubic) / mu_true) * 100
            print("weibull mu error%:         ", mu_error)
    else:
        print("weibull cubic: Cannot compute (too many infinite values)")
    print("-------------------------------------------")
    return

# ------------------------------
# RUN ALL COMPARISONS
# ------------------------------

print("=" * 60)
print("POWER DISTRIBUTION (alpha=2.0)")
print("=" * 60)
compare_power(alpha=2.0)

print("\n" + "=" * 60)
print("LOG-NORMAL DISTRIBUTION (mu=0, sigma=1)")
print("=" * 60)
compare_lognormal(mu=0, sigma=1)

print("\n" + "=" * 60)
print("NORMAL DISTRIBUTION (mu=0, sigma=1)")
print("=" * 60)
compare_normal(mu=0, sigma=1)

print("\n" + "=" * 60)
print("WEIBULL DISTRIBUTION (shape=2.0, scale=1.0)")
print("=" * 60)
compare_weibull(shape=2.0, scale=1.0)
```

```python
import math
import numpy as np
from scipy import stats
from scipy.integrate import quad

# ------------------------------
# Quadratic interpolation functions (reused from previous cell)
# ------------------------------
def lagrange_quadratic(pvals, ovals):
    # Fit quadratic polynomial ax^2 + bx + c through 3 points
    V = np.vstack([pvals**2, pvals, np.ones_like(pvals)]).T
    coeffs = np.linalg.solve(V, ovals)
    return coeffs

def integrate_quadratic(a, b, c, L, R):
    F = lambda x: (a/3)*x**3 + (b/2)*x**2 + c*x
    return F(R) - F(L)

def quadratic_octile_method(octiles):
    O = octiles
    # Use 3 points for quadratic interpolation on each half
    # Left half: use points at 2/8, 3/8, 4/8
    p_left  = np.array([2/8, 3/8, 4/8])
    o_left  = np.array([O[2], O[3], O[4]])
    # Right half: use points at 4/8, 5/8, 7/8
    p_right = np.array([4/8, 5/8, 7/8])
    o_right = np.array([O[4], O[5], O[7]])
    a1, b1, c1 = lagrange_quadratic(p_left, o_left)
    a2, b2, c2 = lagrange_quadratic(p_right, o_right)
    I1 = integrate_quadratic(a1, b1, c1, 0.0, 0.5)
    I2 = integrate_quadratic(a2, b2, c2, 0.5, 1.0)
    mu = I1 + I2
    h = I2 - I1
    return mu, h

def one_median(octiles):
    result = octiles[4]
    return result, 0

# ------------------------------
# OCTILE CALCULATIONS for new distributions
# ------------------------------

def compute_octiles_power(alpha=2.0):
    """
    Power distribution on [0,1]: PDF f(x) = alpha * x^(alpha-1), CDF F(x) = x^alpha
    Quantile function: Q(p) = p^(1/alpha)
    """
    res = [0]
    for i in range(1, 8):
        p = i/8
        res.append(p**(1/alpha))
    return res

def compute_octiles_lognormal(mu=0, sigma=1):
    """
    Log-normal distribution: If X ~ LN(mu, sigma^2), then Q(p) = exp(mu + sigma * Phi^(-1)(p))
    where Phi^(-1) is the inverse standard normal CDF
    """
    res = [0]
    for i in range(1, 8):
        p = i/8
        z = stats.norm.ppf(p, loc=0, scale=1)
        res.append(np.exp(mu + sigma * z))
    return res

def compute_octiles_normal(mu=0, sigma=1):
    """
    Normal distribution: Q(p) = mu + sigma * Phi^(-1)(p)
    For p=0 and p=1, we use very small/large finite values instead of infinity
    """
    res = [0]  # Q(0) - use a very negative finite value
    # Use a very small p value for Q(0) approximation
    z0 = stats.norm.ppf(1e-10, loc=0, scale=1)
    res[0] = mu + sigma * z0
    
    for i in range(1, 8):
        p = i/8
        z = stats.norm.ppf(p, loc=0, scale=1)
        res.append(mu + sigma * z)
    
    # Q(1) - use a very large finite value
    z1 = stats.norm.ppf(1 - 1e-10, loc=0, scale=1)
    res.append(mu + sigma * z1)
    return res

def compute_octiles_weibull(shape=2.0, scale=1.0):
    """
    Weibull distribution: Q(p) = scale * (-log(1-p))^(1/shape)
    """
    res = [0]
    for i in range(1, 8):
        p = i/8
        res.append(scale * ((-math.log(1-p))**(1/shape)))
    res.append(np.inf)  # Q(1) is infinite
    return res

# ------------------------------
# TRUE VALUES CALCULATIONS
# ------------------------------

def compute_true_mu_h_power(alpha):
    """True mu and h for power distribution"""
    mu_true = alpha / (alpha + 1)
    # h = ∫₀.₅¹ p^(1/alpha) dp - ∫₀⁰.⁵ p^(1/alpha) dp
    def Q_power(p):
        return p**(1/alpha)
    I1, _ = quad(Q_power, 0.0, 0.5)
    I2, _ = quad(Q_power, 0.5, 1.0)
    h_true = I2 - I1
    return mu_true, h_true

def compute_true_mu_h_lognormal(mu, sigma):
    """True mu and h for log-normal distribution"""
    mu_true = np.exp(mu + sigma**2/2)
    # h = ∫₀.₅¹ Q(p) dp - ∫₀⁰.⁵ Q(p) dp
    def Q_lognormal(p):
        z = stats.norm.ppf(p, loc=0, scale=1)
        return np.exp(mu + sigma * z)
    I1, _ = quad(Q_lognormal, 0.0, 0.5)
    I2, _ = quad(Q_lognormal, 0.5, 1.0)
    h_true = I2 - I1
    return mu_true, h_true

def compute_true_mu_h_normal(mu, sigma):
    """True mu and h for normal distribution"""
    mu_true = mu
    # h = ∫₀.₅¹ Q(p) dp - ∫₀⁰.⁵ Q(p) dp
    def Q_normal(p):
        z = stats.norm.ppf(p, loc=0, scale=1)
        return mu + sigma * z
    I1, _ = quad(Q_normal, 0.0, 0.5)
    I2, _ = quad(Q_normal, 0.5, 1.0)
    h_true = I2 - I1
    return mu_true, h_true

def compute_true_mu_h_weibull(shape, scale):
    """True mu and h for Weibull distribution"""
    from scipy.special import gamma
    mu_true = scale * gamma(1 + 1/shape)
    # h = ∫₀.₅¹ Q(p) dp - ∫₀⁰.⁵ Q(p) dp
    def Q_weibull(p):
        if p >= 1:
            return np.inf
        return scale * ((-math.log(1-p))**(1/shape))
    I1, _ = quad(Q_weibull, 0.0, 0.5)
    I2, _ = quad(Q_weibull, 0.5, 1.0 - 1e-10)  # Avoid p=1
    h_true = I2 - I1
    return mu_true, h_true

# ------------------------------
# COMPARISON FUNCTIONS
# ------------------------------

def compare_power(alpha=2.0):
    octiles = compute_octiles_power(alpha)
    mu_true, h_true = compute_true_mu_h_power(alpha)
    print("power octiles", octiles)
    print("---------------------------")
    print("power h:                 ", h_true)
    print("power mu:                ", mu_true)
    print("power median:            ", one_median(octiles))
    # Quadratic approximation
    mu_quadratic, h_quadratic = quadratic_octile_method(octiles)
    print("power quadratic h:           ", h_quadratic)
    print("power quadratic mu:          ", mu_quadratic)
    # Error percentages
    if abs(h_true) > 1e-10:
        h_error = abs((h_true - h_quadratic) / h_true) * 100
        print("power h error%:          ", h_error)
    if abs(mu_true) > 1e-10:
        mu_error = abs((mu_true - mu_quadratic) / mu_true) * 100
        print("power mu error%:         ", mu_error)
    print("-------------------------------------------")
    return

def compare_lognormal(mu=0, sigma=1):
    octiles = compute_octiles_lognormal(mu, sigma)
    mu_true, h_true = compute_true_mu_h_lognormal(mu, sigma)
    print("lognormal octiles", octiles)
    print("---------------------------")
    print("lognormal h:                 ", h_true)
    print("lognormal mu:                ", mu_true)
    print("lognormal median:            ", one_median(octiles))
    # Quadratic approximation
    mu_quadratic, h_quadratic = quadratic_octile_method(octiles)
    print("lognormal quadratic h:           ", h_quadratic)
    print("lognormal quadratic mu:          ", mu_quadratic)
    # Error percentages
    if abs(h_true) > 1e-10:
        h_error = abs((h_true - h_quadratic) / h_true) * 100
        print("lognormal h error%:          ", h_error)
    if abs(mu_true) > 1e-10:
        mu_error = abs((mu_true - mu_quadratic) / mu_true) * 100
        print("lognormal mu error%:         ", mu_error)
    print("-------------------------------------------")
    return

def compare_normal(mu=0, sigma=1):
    octiles = compute_octiles_normal(mu, sigma)
    mu_true, h_true = compute_true_mu_h_normal(mu, sigma)
    print("normal octiles", octiles)
    print("---------------------------")
    print("normal h:                 ", h_true)
    print("normal mu:                ", mu_true)
    print("normal median:            ", one_median(octiles))
    # Quadratic approximation
    mu_quadratic, h_quadratic = quadratic_octile_method(octiles)
    print("normal quadratic h:           ", h_quadratic)
    print("normal quadratic mu:          ", mu_quadratic)
    # Error percentages
    if abs(h_true) > 1e-10:
        h_error = abs((h_true - h_quadratic) / h_true) * 100
        print("normal h error%:          ", h_error)
        if abs(mu_true) > 1e-10:
            mu_error = abs((mu_true - mu_quadratic) / mu_true) * 100
            print("normal mu error%:         ", mu_error)
    else:
        # When mu_true is 0, print absolute error instead
            mu_abs_error = abs(mu_true - mu_quadratic)
            print("normal mu absolute error:  ", mu_abs_error)
            print("normal mu error%:         N/A (mu_true = 0)")
    print("-------------------------------------------")
    return

def compare_weibull(shape=2.0, scale=1.0):
    octiles = compute_octiles_weibull(shape, scale)
    mu_true, h_true = compute_true_mu_h_weibull(shape, scale)
    print("weibull octiles", octiles)
    print("---------------------------")
    print("weibull h:                 ", h_true)
    print("weibull mu:                ", mu_true)
    print("weibull median:            ", one_median(octiles))
    # Quadratic approximation (only use finite octiles)
    octiles_finite = [q for q in octiles if np.isfinite(q)]
    if len(octiles_finite) >= 8:
        mu_quadratic, h_quadratic = quadratic_octile_method(octiles_finite)
        print("weibull quadratic h:           ", h_quadratic)
        print("weibull quadratic mu:          ", mu_quadratic)
        # Error percentages
        if abs(h_true) > 1e-10:
            h_error = abs((h_true - h_quadratic) / h_true) * 100
            print("weibull h error%:          ", h_error)
        if abs(mu_true) > 1e-10:
            mu_error = abs((mu_true - mu_quadratic) / mu_true) * 100
            print("weibull mu error%:         ", mu_error)
    else:
        print("weibull quadratic: Cannot compute (too many infinite values)")
    print("-------------------------------------------")
    return

# ------------------------------
# RUN ALL COMPARISONS
# ------------------------------

print("=" * 60)
print("POWER DISTRIBUTION (alpha=2.0)")
print("=" * 60)
compare_power(alpha=2.0)

print("\n" + "=" * 60)
print("LOG-NORMAL DISTRIBUTION (mu=0, sigma=1)")
print("=" * 60)
compare_lognormal(mu=0, sigma=1)

print("\n" + "=" * 60)
print("NORMAL DISTRIBUTION (mu=0, sigma=1)")
print("=" * 60)
compare_normal(mu=0, sigma=1)

print("\n" + "=" * 60)
print("WEIBULL DISTRIBUTION (shape=2.0, scale=1.0)")
print("=" * 60)
compare_weibull(shape=2.0, scale=1.0)
```

```python
import math
import numpy as np

# ------------------------------
# Quadratic endpoint estimation with trapezoid approximation
# ------------------------------
def quadratic_endpoint_trapezoid_method(octiles):
    """
    Uses quadratic interpolation to estimate endpoints, then trapezoid approximation.
    C_L(0) = 3O_1 - 3O_2 + O_3
    C_R(1) = O_5 - 3O_6 + 3O_7
    I1 = (5O_1 - O_2 + 3O_3 + O_4) / 16
    I2 = (O_4 + 3O_5 - O_6 + 5O_7) / 16
    mu = I1 + I2
    h = I2 - I1
    """
    O = octiles
    O_1, O_2, O_3, O_4 = O[1], O[2], O[3], O[4]
    O_5, O_6, O_7 = O[5], O[6], O[7]
    
    # Estimate endpoints using quadratic interpolation
    C_L_0 = 3*O_1 - 3*O_2 + O_3
    C_R_1 = O_5 - 3*O_6 + 3*O_7
    
    # Trapezoid approximation with estimated endpoints
    I1 = (5*O_1 - O_2 + 3*O_3 + O_4) / 16.0
    I2 = (O_4 + 3*O_5 - O_6 + 5*O_7) / 16.0
    
    mu = I1 + I2
    h = I2 - I1
    
    return mu, h

def one_median(octiles):
    result = octiles[4]
    return result, 0

def compute_octiles_uniform():
    res = [0]
    for i in range(1,8):
        p = i/8
        res.append(p)
    return res

def compute_octiles_exponential(my_lambda=1):
    res = [0]
    for i in range(1,8):
        p = i/8
        res.append((-math.log(1-p))/my_lambda)
    return res    

def compute_octiles_laplace(my_mu=0, my_lambda=1):
    res = [0]
    O_1 = my_mu + my_lambda*math.log(2*1/8)
    O_2 = my_mu + my_lambda*math.log(2*2/8)
    O_3 = my_mu + my_lambda*math.log(2*3/8) 
    O_4 = my_mu + my_lambda*math.log(2*4/8)
    O_5 = my_mu - my_lambda*math.log(2 - 2*5/8)
    O_6 = my_mu - my_lambda*math.log(2 - 2*6/8)
    O_7 = my_mu - my_lambda*math.log(2 - 2*7/8)     
    res =[0, O_1, O_2, O_3, O_4, O_5, O_6, O_7]
    return res    

def compute_octiles_pareto(my_alpha = 1.5, my_beta = 1):
    res = [0]
    for i in range(1,8):
        p = i/8
        res.append(my_beta *((1-p)**(-1/my_alpha)))
    return res    

def compare_pareto(my_alpha=1.5, my_beta=1):
    octiles = compute_octiles_pareto(my_alpha, my_beta)
    print("pareto octiles", octiles)
    my_h = my_beta * my_alpha *((2**(1/my_alpha))-1)/(my_alpha -1)
    my_mu = my_alpha * my_beta /(my_alpha -1)
    print("---------------------------")
    print("pareto h:                 ", my_h)
    print("pareto mu:                ", my_mu)
    print("pareto median:            ", one_median(octiles))
    # Quadratic endpoint trapezoid approximation
    mu_quad_trap, h_quad_trap = quadratic_endpoint_trapezoid_method(octiles)
    print("pareto quadratic endpoint trapezoid h:           ", h_quad_trap)
    print("pareto quadratic endpoint trapezoid mu:          ", mu_quad_trap)
    # Error percentages
    if abs(my_h) > 1e-10:
        h_error = abs((my_h - h_quad_trap) / my_h) * 100
        print("pareto h error%:          ", h_error)
    if abs(my_mu) > 1e-10:
        mu_error = abs((my_mu - mu_quad_trap) / my_mu) * 100
        print("pareto mu error%:         ", mu_error)
    print("-------------------------------------------")
    return
 
def compare_laplace(my_mu=0, my_lambda=1):
    octiles = compute_octiles_laplace(my_mu, my_lambda)
    print("laplace octiles", octiles)
    h = my_lambda/math.sqrt(2)
    mu = 0
    dist = "laplace "
    print("---------------------------")
    print(dist + "h:                 ", h)
    print(dist + "mu:                ", mu)
    print(dist + "median:            ", one_median(octiles))
    # Quadratic endpoint trapezoid approximation
    mu_quad_trap, h_quad_trap = quadratic_endpoint_trapezoid_method(octiles)
    print(dist + "quadratic endpoint trapezoid h:           ", h_quad_trap)
    print(dist + "quadratic endpoint trapezoid mu:          ", mu_quad_trap)
    # Error percentages
    if abs(h) > 1e-10:
        h_error = abs((h - h_quad_trap) / h) * 100
        print(dist + "h error%:          ", h_error)
    if abs(mu) > 1e-10:
        mu_error = abs((mu - mu_quad_trap) / mu) * 100
        print(dist + "mu error%:         ", mu_error)
    else:
        print(dist + "mu error%:         0")
    print("----------------------------")
    return   
    
def compare_exponential(my_lambda=1):
    octiles = compute_octiles_exponential(my_lambda)
    print("exponential octiles", octiles)
    h = math.log(2)/my_lambda
    mu = 1/my_lambda
    print("---------------------------")
    print("exp h:                 ", h)
    print("exp mu:                ", mu)
    print("exp median:            ", one_median(octiles))
    # Quadratic endpoint trapezoid approximation
    mu_quad_trap, h_quad_trap = quadratic_endpoint_trapezoid_method(octiles)
    print("exp quadratic endpoint trapezoid h:           ", h_quad_trap)
    print("exp quadratic endpoint trapezoid mu:          ", mu_quad_trap)
    # Error percentages
    if abs(h) > 1e-10:
        h_error = abs((h - h_quad_trap) / h) * 100
        print("exp h error%:           ", h_error)
    if abs(mu) > 1e-10:
        mu_error = abs((mu - mu_quad_trap) / mu) * 100
        print("exp mu error%:          ", mu_error)
    print("----------------------------")
    return

def compare_uniform():
    octiles = compute_octiles_uniform()
    print("uniform octiles", octiles)
    mu, h = 1/2, 1/4
    print("---------------------------")
    print("uniform h:                 ", h)
    print("uniform mu:                ", mu)
    print("uniform median:            ", one_median(octiles))
    # Quadratic endpoint trapezoid approximation
    mu_quad_trap, h_quad_trap = quadratic_endpoint_trapezoid_method(octiles)
    print("uniform quadratic endpoint trapezoid h:           ", h_quad_trap)
    print("uniform quadratic endpoint trapezoid mu:          ", mu_quad_trap)
    # Error percentages
    if abs(h) > 1e-10:
        h_error = abs((h - h_quad_trap) / h) * 100
        print("uniform h error%:         ", h_error)
    if abs(mu) > 1e-10:
        mu_error = abs((mu - mu_quad_trap) / mu) * 100
        print("uniform mu error%:        ", mu_error)
    print("----------------------------")
    return
    
# Run comparisons with similar parameters
compare_laplace(0, 1) 
compare_exponential(1)
compare_uniform()
compare_pareto(2.0, 1/2)
```

```python
import math
import numpy as np

# Define compute_octiles_exponential function
def compute_octiles_exponential(my_lambda=1):
    res = [0]
    for i in range(1,8):
        p = i/8
        res.append((-math.log(1-p))/my_lambda)
    return res

# Define quadratic interpolation functions
def lagrange_quadratic(pvals, ovals):
    # Fit quadratic polynomial ax^2 + bx + c through 3 points
    V = np.vstack([pvals**2, pvals, np.ones_like(pvals)]).T
    coeffs = np.linalg.solve(V, ovals)
    return coeffs

def integrate_quadratic(a, b, c, L, R):
    F = lambda x: (a/3)*x**3 + (b/2)*x**2 + c*x
    return F(R) - F(L)

def quadratic_octile_method(octiles):
    O = octiles
    # Use 3 points for quadratic interpolation on each half
    # Left half: use points at 2/8, 3/8, 4/8
    p_left  = np.array([2/8, 3/8, 4/8])
    o_left  = np.array([O[2], O[3], O[4]])
    # Right half: use points at 4/8, 5/8, 7/8
    p_right = np.array([4/8, 5/8, 7/8])
    o_right = np.array([O[4], O[5], O[7]])
    a1, b1, c1 = lagrange_quadratic(p_left, o_left)
    a2, b2, c2 = lagrange_quadratic(p_right, o_right)
    I1 = integrate_quadratic(a1, b1, c1, 0.0, 0.5)
    I2 = integrate_quadratic(a2, b2, c2, 0.5, 1.0)
    mu = I1 + I2
    h = I2 - I1
    return mu, h

# Use theoretical octiles (with true lambda = 1.0)
true_lambda = 1.0
theoretical_octiles = compute_octiles_exponential(true_lambda)

# Extract octiles: O1, O2, O3, O4 (median), O5, O6, O7
O1, O2, O3, O4, O5, O6, O7 = theoretical_octiles[1], theoretical_octiles[2], theoretical_octiles[3], theoretical_octiles[4], theoretical_octiles[5], theoretical_octiles[6], theoretical_octiles[7]

# Compute H using equation 4.9: H ≈ [5(O7-O1) + 3(O5-O3) - (O6-O2)] / 16
H = (5 * (O7 - O1) + 3 * (O5 - O3) - (O6 - O2)) / 16

# Compute λ₁ = log(2)/H (eq. 4.9 method)
lambda_1 = math.log(2) / H

# Compute λ₂ = log(2)/median (median/MOM method)
median = O4
lambda_2 = math.log(2) / median

# Print results
print("Using THEORETICAL octiles (with true λ = 1.0):")
print(f"Theoretical octiles: {theoretical_octiles}")
print()
print(f"O1 = {O1:.6f}")
print(f"O2 = {O2:.6f}")
print(f"O3 = {O3:.6f}")
print(f"O4 (median) = {O4:.6f}")
print(f"O5 = {O5:.6f}")
print(f"O6 = {O6:.6f}")
print(f"O7 = {O7:.6f}")
print()
print(f"H (from eq. 4.9) = {H:.6f}")
print()
print(f"λ₁ (eq. 4.9 method) = log(2)/H = {lambda_1:.6f}")
print(f"λ₂ (median/MOM method) = log(2)/median = {lambda_2:.6f}")
print()
print(f"True λ = {true_lambda:.6f}")
print()
print("Errors:")
print(f"Error in λ₁: {abs(lambda_1 - true_lambda):.6f} ({abs((lambda_1 - true_lambda)/true_lambda)*100:.2f}%)")
print(f"Error in λ₂: {abs(lambda_2 - true_lambda):.6f} ({abs((lambda_2 - true_lambda)/true_lambda)*100:.2f}%)")
```

```python
import math
import numpy as np

# -------------------------------
# Functions
# -------------------------------
def compute_octiles_exponential(my_lambda=1):
    res = [0]
    for i in range(1,8):
        p = i/8
        res.append((-math.log(1-p))/my_lambda)
    return res

def lagrange_quadratic(pvals, ovals):
    V = np.vstack([pvals**2, pvals, np.ones_like(pvals)]).T
    coeffs = np.linalg.solve(V, ovals)
    return coeffs

def integrate_quadratic(a, b, c, L, R):
    F = lambda x: (a/3)*x**3 + (b/2)*x**2 + c*x
    return F(R) - F(L)

def quadratic_octile_method(octiles):
    O = octiles
    p_left  = np.array([2/8, 3/8, 4/8])
    o_left  = np.array([O[2], O[3], O[4]])
    p_right = np.array([4/8, 5/8, 7/8])
    o_right = np.array([O[4], O[5], O[7]])
    a1, b1, c1 = lagrange_quadratic(p_left, o_left)
    a2, b2, c2 = lagrange_quadratic(p_right, o_right)
    I1 = integrate_quadratic(a1, b1, c1, 0.0, 0.5)
    I2 = integrate_quadratic(a2, b2, c2, 0.5, 1.0)
    mu = I1 + I2
    h = I2 - I1
    return mu, h

# -------------------------------
# True values
# -------------------------------
true_lambda = 1.0
true_median = math.log(2)/true_lambda

# Theoretical octiles
octiles = compute_octiles_exponential(true_lambda)
O1, O2, O3, O4, O5, O6, O7 = octiles[1:8]

# H and lambda via eq 4.9
H = (5*(O7-O1) + 3*(O5-O3) - (O6-O2)) / 16
lambda_eq49 = math.log(2)/H
lambda_median = math.log(2)/O4

# -------------------------------
# Simulation: estimate lambda from samples
# -------------------------------
rng = np.random.default_rng(42)
n_samples = 50
n_sim = 2000

lambda_estimates = []
median_estimates = []

for _ in range(n_sim):
    sample = rng.exponential(scale=1/true_lambda, size=n_samples)
    lam_hat = 1/np.mean(sample)  # MLE for lambda
    lambda_estimates.append(lam_hat)
    median_estimates.append(math.log(2)/lam_hat)

lambda_estimates = np.array(lambda_estimates)
median_estimates = np.array(median_estimates)

# Three median estimators
# 1. ln2 / mean(lambda)
m_from_mean_lambda = math.log(2)/np.mean(lambda_estimates)
# 2. mean of median estimates
m_from_mean_median = np.mean(median_estimates)
# 3. median of median estimates
m_from_median_median = np.median(median_estimates)

# Errors
def report_error(est, true_val):
    abs_err = est - true_val
    pct_err = abs_err/true_val*100
    return abs_err, pct_err

err_eq49, pct_eq49 = report_error(lambda_eq49, true_lambda)
err_median, pct_median = report_error(lambda_median, true_lambda)
err_mean_lambda, pct_mean_lambda = report_error(math.log(2)/np.mean(lambda_estimates), true_median)
err_mean_median, pct_mean_median = report_error(m_from_mean_median, true_median)
err_median_median, pct_median_median = report_error(m_from_median_median, true_median)

# -------------------------------
# Print results
# -------------------------------
print("THEORETICAL OCTILES:")
print(octiles)
print(f"H = {H:.6f}")
print(f"Lambda eq4.9 = {lambda_eq49:.6f}, error = {err_eq49:.6f} ({pct_eq49:.2f}%)")
print(f"Lambda median method = {lambda_median:.6f}, error = {err_median:.6f} ({pct_median:.2f}%)")
print()
print("SIMULATION-BASED MEDIAN ESTIMATORS:")
print(f"1) ln2 / mean(lambda) = {m_from_mean_lambda:.6f}, error = {err_mean_lambda:.6f} ({pct_mean_lambda:.2f}%)")
print(f"2) mean of median estimates = {m_from_mean_median:.6f}, error = {err_mean_median:.6f} ({pct_mean_median:.2f}%)")
print(f"3) median of median estimates = {m_from_median_median:.6f}, error = {err_median_median:.6f} ({pct_median_median:.2f}%)")
print()
print(f"True lambda = {true_lambda}, True median = {true_median:.6f}")
```

```python
import math
import numpy as np
import pandas as pd
from scipy import stats

def compute_octiles_from_data(data):
    """
    Compute octiles from real data.
    
    Parameters:
    -----------
    data : array-like
        One-dimensional array of numerical values
        
    Returns:
    --------
    octiles : list
        List of 8 octiles [O0, O1, O2, O3, O4, O5, O6, O7]
        where O4 is the median (50th percentile)
    """
    data = np.array(data)
    data = data[~np.isnan(data)]  # Remove NaN values
    
    if len(data) < 8:
        raise ValueError(f"Need at least 8 data points, got {len(data)}")
    
    percentiles = [0, 12.5, 25.0, 37.5, 50.0, 62.5, 75.0, 87.5]
    octiles = np.percentile(data, percentiles).tolist()
    
    # O[0] should be minimum (or a very small value)
    octiles[0] = np.min(data)
    
    return octiles


def apply_all_approximations(octiles):
    """
    Apply all approximation methods to compute mean (μ) and MAD (H) from octiles.
    
    Parameters:
    -----------
    octiles : list or array
        List of 8 octiles [O0, O1, O2, O3, O4, O5, O6, O7]
        
    Returns:
    --------
    results : dict
        Dictionary containing results for all methods
        Each method returns (mu, h) tuple where available
    """
    results = {}
    
    # Ensure octiles is a list with proper indexing
    if isinstance(octiles, np.ndarray):
        octiles = octiles.tolist()
    
    # Basic methods
    try:
        results['one_median'] = one_median(octiles)
    except:
        results['one_median'] = (None, None)
    
    try:
        results['two_quartiles'] = two_quartiles(octiles)
    except:
        results['two_quartiles'] = (None, None)
    
    try:
        results['four_quartiles'] = four_quartiles(octiles)
    except:
        results['four_quartiles'] = (None, None)
    
    try:
        results['four_octiles'] = four_octiles(octiles)
    except:
        results['four_octiles'] = (None, None)
    
    try:
        results['all_octiles'] = all_octiles(octiles)
    except:
        results['all_octiles'] = (None, None)
    
    # Trapezoid methods
    try:
        mu_trap = trapezoid(octiles)
        results['trapezoid'] = (mu_trap, None)
    except:
        results['trapezoid'] = (None, None)
    
    try:
        mu_2r2t = two_rect_two_trap(octiles)
        results['two_rect_two_trap'] = (mu_2r2t, None)
    except:
        results['two_rect_two_trap'] = (None, None)
    
    try:
        mu_2r1t = two_rect_one_trap(octiles)
        results['two_rect_one_trap'] = (mu_2r1t, None)
    except:
        results['two_rect_one_trap'] = (None, None)
    
    # Simpson methods
    try:
        mu_simp = simpson(octiles)
        results['simpson'] = (mu_simp, None)
    except:
        results['simpson'] = (None, None)
    
    try:
        mu_simp3 = simpson_3(octiles)
        results['simpson_3'] = (mu_simp3, None)
    except:
        results['simpson_3'] = (None, None)
    
    try:
        mu_simpc = simpson_composite(octiles)
        results['simpson_composite'] = (mu_simpc, None)
    except:
        results['simpson_composite'] = (None, None)
    
    # Special methods that return both mu and h
    try:
        results['one_rect_three_trapezoids'] = one_rect_three_trapezoids(octiles)
    except:
        results['one_rect_three_trapezoids'] = (None, None)
    
    try:
        results['one_triangle_three_trapezoids'] = one_triangle_three_trapezoids(octiles)
    except:
        results['one_triangle_three_trapezoids'] = (None, None)
    
    # Cubic interpolation method (if available)
    try:
        if 'cubic_octile_method' in globals():
            results['cubic'] = cubic_octile_method(octiles)
        else:
            results['cubic'] = (None, None)
    except:
        results['cubic'] = (None, None)
    
    # Quadratic interpolation method (if available)
    try:
        if 'quadratic_octile_method' in globals():
            results['quadratic'] = quadratic_octile_method(octiles)
        else:
            results['quadratic'] = (None, None)
    except:
        results['quadratic'] = (None, None)
    
    # Quadratic endpoint trapezoid method (if available)
    try:
        if 'quadratic_endpoint_trapezoid_method' in globals():
            results['quadratic_endpoint_trapezoid'] = quadratic_endpoint_trapezoid_method(octiles)
        else:
            results['quadratic_endpoint_trapezoid'] = (None, None)
    except:
        results['quadratic_endpoint_trapezoid'] = (None, None)
    
    return results


def analyze_real_dataset(data, dataset_name="Dataset", compute_true_values=True):
    """
    Complete analysis of a real dataset using all approximation methods.
    
    Parameters:
    -----------
    data : array-like
        One-dimensional array of numerical values
    dataset_name : str
        Name of the dataset for reporting
    compute_true_values : bool
        Whether to compute true mean and MAD from data (for comparison)
        
    Returns:
    --------
    report : dict
        Comprehensive report with octiles, approximations, and true values
    """
    data = np.array(data)
    data = data[~np.isnan(data)]
    
    if len(data) < 8:
        raise ValueError(f"Need at least 8 data points, got {len(data)}")
    
    # Compute octiles
    octiles = compute_octiles_from_data(data)
    
    # Apply all approximations
    approximations = apply_all_approximations(octiles)
    
    # Compute true values from data (if requested)
    true_mu = np.mean(data)
    true_h = np.mean(np.abs(data - np.median(data)))  # True MAD
    
    report = {
        'dataset_name': dataset_name,
        'sample_size': len(data),
        'octiles': octiles,
        'octiles_dict': {
            'O1': octiles[1], 'O2': octiles[2], 'O3': octiles[3],
            'O4 (median)': octiles[4], 'O5': octiles[5], 
            'O6': octiles[6], 'O7': octiles[7]
        },
        'true_mean': true_mu,
        'true_mad': true_h,
        'approximations': approximations
    }
    
    return report


def print_comparison_table(report, include_errors=True):
    """
    Print a formatted comparison table of all methods.
    
    Parameters:
    -----------
    report : dict
        Report dictionary from analyze_real_dataset
    include_errors : bool
        Whether to include error percentages compared to true values
    """
    print("=" * 100)
    print(f"ANALYSIS: {report['dataset_name'].upper()}")
    print("=" * 100)
    print(f"Sample size: {report['sample_size']}")
    print(f"True Mean (μ): {report['true_mean']:.6f}")
    print(f"True MAD (H): {report['true_mad']:.6f}")
    print()
    
    print("Octiles:")
    for key, value in report['octiles_dict'].items():
        print(f"  {key}: {value:.6f}")
    print()
    
    print("=" * 100)
    print("APPROXIMATION RESULTS")
    print("=" * 100)
    
    if include_errors:
        header = f"{'Method':<35} {'μ':<15} {'μ Error %':<15} {'H (MAD)':<15} {'H Error %':<15}"
    else:
        header = f"{'Method':<35} {'μ':<15} {'H (MAD)':<15}"
    
    print(header)
    print("-" * 100)
    
    true_mu = report['true_mean']
    true_h = report['true_mad']
    
    for method_name, (mu, h) in sorted(report['approximations'].items()):
        if mu is None and h is None:
            continue
            
        if include_errors:
            # Calculate errors
            if mu is not None and abs(true_mu) > 1e-10:
                mu_error = abs((mu - true_mu) / true_mu) * 100
            elif mu is not None:
                mu_error = abs(mu - true_mu)
            else:
                mu_error = None
            
            if h is not None and abs(true_h) > 1e-10:
                h_error = abs((h - true_h) / true_h) * 100
            elif h is not None:
                h_error = abs(h - true_h)
            else:
                h_error = None
            
            mu_str = f"{mu:.6f}" if mu is not None else "N/A"
            mu_err_str = f"{mu_error:.4f}" if mu_error is not None else "N/A"
            h_str = f"{h:.6f}" if h is not None else "N/A"
            h_err_str = f"{h_error:.4f}" if h_error is not None else "N/A"
            
            print(f"{method_name:<35} {mu_str:<15} {mu_err_str:<15} {h_str:<15} {h_err_str:<15}")
        else:
            mu_str = f"{mu:.6f}" if mu is not None else "N/A"
            h_str = f"{h:.6f}" if h is not None else "N/A"
            print(f"{method_name:<35} {mu_str:<15} {h_str:<15}")
    
    print()


# Example usage:
print("Functions loaded successfully!")
print("\nUsage:")
print("  data = your_data_array  # 1D numpy array or list")
print("  report = analyze_real_dataset(data, dataset_name='My Dataset')")
print("  print_comparison_table(report)")
```
