```python
import math

a = 3* math.log(3)/(2*math.log(2))-2
print(a)
b=math.log(2)/(5*math.log(2)-3*math.log(3))
print(b)


beta = 1
alpha = 1.5

def trapezoid(octiles):
    O_1, O_2, O_3 = octiles[1], octiles[2], octiles[3]
    O_4, O_5, O_6, O_7 = octiles[4], octiles[5], octiles[6], octiles[7]
    first_half = O_1/8 + (O_1 + O_2)/16 + (O_2 + O_3)/16 + (O_3 + O_4)/16 
    second_half = (O_4+O_5)/16 +(O_5+O_6)/16 + (O_6 + O_7)/16 + O_7/8
    result = first_half + second_half
    return result

def simpson(octiles):
    O_1, O_2, O_3 = octiles[1], octiles[2], octiles[3]
    O_4, O_5, O_6, O_7 = octiles[4], octiles[5], octiles[6], octiles[7]
    first_half = O_1/4 + (O_2 + 4*O_3 + O_4)/24
    second_half = (O_4 + 4*O_5 + O_6)/24 + O_7/4
    result = first_half + second_half
    return result

def simpson_3(octiles):
    O_1, O_2, O_3 = octiles[1], octiles[2], octiles[3]
    O_4, O_5, O_6, O_7 = octiles[4], octiles[5], octiles[6], octiles[7]
    left_part = (O_1 + 4*O_2 + O_3)/24
    middle_part = (O_3 + 4*O_4 + O_5)/24
    right_part = (O_5 + 4*O_6 + O_7)/24
    result = O_1/8 + left_part + middle_part + right_part + O_7/8
    return result

def simpson_composite(octiles):
    O_1, O_2, O_3 = octiles[1], octiles[2], octiles[3]
    O_4, O_5, O_6, O_7 = octiles[4], octiles[5], octiles[6], octiles[7]
    first_half = O_1/8 + (O_1 + 3*O_2 + 3*O_3 + O_4)*3/64
    second_half = (O_4 + 3*O_5 + 3*O_6 + O_7)*3/64 + O_7/8
    result = first_half + second_half
    return result

def one_median(octiles):
    result = octiles[4]
    return result, 0

def two_quartiles(octiles):
    O_2 = octiles[2]
    O_6 = octiles[6]
    mu_two = (O_2 + O_6)/2
    h_two = (O_6 - O_2)/2
    return mu_two, h_two

def four_quartiles(octiles):
    O_2 = octiles[2]
    O_4 = octiles[4]
    O_6 = octiles[6]
    mu_four = (O_2 + 2*O_4 + O_4)/4
    h_four  = (O_6 - O_2)/2    
    return mu_four, h_four

def four_octiles(octiles):
    O_1 = octiles[1]
    O_3 = octiles[3]
    O_5 = octiles[5]
    O_7 = octiles[7]
    mu_four = (O_1 + O_3 + O_5 + O_7)/4
    h_four  = ((O_5 + O_7)-(O_1+O_3))/4   
    return mu_four, h_four

def all_octiles(octiles):
    O_1, O_2, O_3 = octiles[1], octiles[2], octiles[3]
    O_4, O_5, O_6, O_7 = octiles[4], octiles[5], octiles[6], octiles[7]
    first_half  = (O_1 + O_2 + O_3 + O_4)/8
    second_half = (O_4 + O_5 + O_6 + O_7)/8
    h = second_half - first_half
    result = first_half + second_half, h
    
    return result

def two_rect_two_trap(octiles):
    O_1, O_2, O_3 = octiles[1], octiles[2], octiles[3]
    O_4, O_5, O_6, O_7 = octiles[4], octiles[5], octiles[6], octiles[7]
    first_half  = 5*O_1/16 +3*O_4/16
    second_half = 3*O_4/16 + 5*O_7/16
    result = first_half + second_half
    return result

def two_rect_one_trap(octiles):
    O_1, O_2, O_3 = octiles[1], octiles[2], octiles[3]
    O_4, O_5, O_6, O_7 = octiles[4], octiles[5], octiles[6], octiles[7]
    result = (O_1 + O_7)/2 
    return result

def one_rect_three_trapezoids(octiles):
    O_1, O_2, O_3 = octiles[1], octiles[2], octiles[3]
    O_4, O_5, O_6, O_7 = octiles[4], octiles[5], octiles[6], octiles[7]
    I_1 = (2*O_1 + 2*O_2 +2*O_3 + O_4)/16.0
    I_2 =(O_4 + 2*O_5 + 2*O_6 + 3*O_7)/16.0
    new_mu = I_1 + I_2
    new_h  = I_2 - I_1
    return new_mu, new_h


def one_triangle_three_trapezoids(octiles):
    O_1, O_2, O_3 = octiles[1], octiles[2], octiles[3]
    O_4, O_5, O_6, O_7 = octiles[4], octiles[5], octiles[6], octiles[7]
    I_1 = (3*O_1 + 2*O_2 +2*O_3 + O_4)/16.0
    I_2 =(O_4 + 2*O_5 + O_6 + 4*O_7)/16.0
    new_mu = I_1 + I_2
    new_h  = I_2 - I_1
    return new_mu, new_h

def four_trapezoids(octiles):
    O_1, O_2, O_3 = octiles[1], octiles[2], octiles[3]
    O_4, O_5, O_6, O_7 = octiles[4], octiles[5], octiles[6], octiles[7]
    I_1 = (5*O_1 - O_2 +3*O_3 + O_4)/16.0
    I_2 =(O_4 + 3*O_5 - O_6 + 5*O_7)/16.0
    new_mu = I_1 + I_2
    new_h  = I_2 - I_1
    return new_mu, new_h
   



def compute_octiles_uniform():
    res = [0]
    for i in range(1,8):
        p = i/8
        res.append(p)
    return res

def compute_octiles_exponential(my_rate=1):
    res = [0]
    for i in range(1,8):
        p = i/8
        res.append((-math.log(1-p))/my_rate)
    return res    

def compute_octiles_laplace(my_mu=0, my_b=1):
    res = [0]
    O_1 = my_mu + my_b*math.log(2*1/8)
    O_2 = my_mu + my_b*math.log(2*2/8)
    O_3 = my_mu + my_b*math.log(2*3/8) 
    O_4 = my_mu + my_b*math.log(2*4/8)
    O_5 = my_mu - my_b*math.log(2 - 2*5/8)
    O_6 = my_mu - my_b*math.log(2 - 2*6/8)
    O_7 = my_mu - my_b*math.log(2 - 2*7/8)     
    res =[0, O_1, O_2, O_3, O_4, O_5, O_6, O_7]
    return res    

def compute_octiles_pareto(my_alpha = 1.5, my_beta = 1):
    res = [0]
    for i in range(1,8):
        p = i/8
        res.append(my_beta *((1-p)**(-1/my_alpha)))
    return res    


def compare_pareto(my_alpha=1.5, my_beta=1):
    octiles = compute_octiles_pareto(my_alpha, my_beta)
    print("pareto octiles", octiles)
    O_1, O_2, O_3 = octiles[1], octiles[2], octiles[3]
    O_4, O_5, O_6, O_7 = octiles[4], octiles[5], octiles[6], octiles[7]
    my_h = my_beta * my_alpha *((2**(1/my_alpha))-1)/(my_alpha -1)
    my_mu = my_alpha * my_beta /(my_alpha -1)
    dist="Pareto"
    print("pareto h:                 ", my_h)
    print("pareto mu:                ", my_mu)
    print("pareto median:            ", one_median(octiles))    
    print("pareto 2-quartiles:       ", two_quartiles(octiles))    
    print("pareto 4-quartiles        ", four_quartiles(octiles))    
    print("pareto 4-octiles :        ",four_octiles(octiles))   
    print("pareto all-octiles :      ", all_octiles(octiles))
    print("pareto trap mu:           ", trapezoid(octiles))
    print("pareto 2-rec-2-trap       ", two_rect_two_trap(octiles))
    print("pareto 2-rec-1-trap       ", two_rect_one_trap(octiles))
    print("pareto simpson:           ", simpson(octiles))
    print("pareto simpson_3:         ", simpson_3(octiles))
    print("pareto simpson_composite: ", simpson_composite(octiles))
    print("pareto 1 rec 3 trap mu: ", one_rect_three_trapezoids(octiles)[0])
    print("pareto 1 rec 3 trap h: ", one_rect_three_trapezoids(octiles)[1])
    print("pareto 1 tri 3 trap mu: ", one_triangle_three_trapezoids(octiles)[0])
    print("pareto 1 tri 3 trap h: ", one_triangle_three_trapezoids(octiles)[1])
    print("pareto 4 trapezoids: ", four_trapezoids(octiles)[0])
    print("pareto 4 trapezoids ", four_trapezoids(octiles)[1])
    print("-------------------------------------------")
    return
 
def compare_laplace(my_mu=0, my_b=1):
    """my_b is the Laplace scale parameter b used in compute_octiles_laplace.
    True MAD about the median is H=E|X-M|=b, so report h=my_b.
    Paper-style variance-one Laplace: Var=2b^2 => b=1/sqrt(2), call compare_laplace(0, 1/math.sqrt(2)). Unit scale b=1: compare_laplace(0, 1)."""
    octiles = compute_octiles_laplace(my_mu, my_b)
    print("laplace octiles", octiles)
    h = my_b  # MAD about the median: E|X-M| = scale b for Laplace(mu, b)
    mu = my_mu
    dist = " Laplace "
    print("---------------------------")
    print(dist + "parameter type:      b (scale)")
    print(dist + "h:                 ", h)
    print(dist + " mu:                ", mu)
    print(dist + " median:            ", one_median(octiles))    
    print(dist + " 2-quartiles:       ", two_quartiles(octiles))    
    print(dist + " 4-quartiles        ", four_quartiles(octiles))    
    print(dist + " 4-octiles :        ", four_octiles(octiles))   
    print(dist + " all-octiles :      ", all_octiles(octiles))
    print(dist + " trap mu:           ", trapezoid(octiles))
    print(dist + " 2-rec-2-trap       ", two_rect_two_trap(octiles))
    print(dist + " 2-rec-1-trap       ", two_rect_one_trap(octiles))    
    print(dist + " simpson:           ", simpson(octiles))
    print(dist + " simpson_3:         ", simpson_3(octiles))
    print(dist + " simpson_composite: ", simpson_composite(octiles))
    print(dist + " 1 rec 3 trap mu: ", one_rect_three_trapezoids(octiles)[0])
    print(dist + " 1 rec 3 trap h: ", one_rect_three_trapezoids(octiles)[1])
    print(dist + " 1 tri 3 trap mu: ", one_triangle_three_trapezoids(octiles)[0])
    print(dist + " 1 tri 3 trap h: ",  one_triangle_three_trapezoids(octiles)[1])
    print(dist + " 4 trapezoids mu: ", four_trapezoids(octiles)[0])
    print(dist + " 4 trapezoids h: ", four_trapezoids(octiles)[1])
    print("----------------------------")
    return   
    
    
    
def compare_exponential(my_rate=1):
    octiles = compute_octiles_exponential(my_rate)
    print("exponential octiles", octiles)
    h = math.log(2)/my_rate
    mu = 1/my_rate
    dist ="exp "
    print("---------------------------")
    print("exp parameter type:     lambda (rate)")
    print("exp h:                 ", h)
    print("exp mu:                ", mu)
    print("exp median:            ", one_median(octiles))    
    print("exp 2-quartiles:       ", two_quartiles(octiles))    
    print("exp 4-quartiles        ", four_quartiles(octiles))    
    print("exp 4-octiles :        ", four_octiles(octiles))   
    print("exp all-octiles :      ", all_octiles(octiles))
    print("exp trap mu:           ", trapezoid(octiles))
    print("exp 2-rec-2-trap       ", two_rect_two_trap(octiles))
    print("exp 2-rec-1-trap       ", two_rect_one_trap(octiles))    
    print("exp simpson:           ", simpson(octiles))
    print("exp simpson_3:         ", simpson_3(octiles))
    print("exp simpson_composite: ", simpson_composite(octiles))
    print("exp 1 rec 3 trap mu: ", one_rect_three_trapezoids(octiles)[0])
    print("exp 1 rec 3 trap h: ", one_rect_three_trapezoids(octiles)[1])
    print(dist + " 1 tri 3 trap mu: ", one_triangle_three_trapezoids(octiles)[0])
    print(dist + " 1 tri 3 trap h: ",  one_triangle_three_trapezoids(octiles)[1])
    print(dist + " 4 trapezoids mu: ", four_trapezoids(octiles)[0])
    print(dist + " 4 trapezoids h: ", four_trapezoids(octiles)[1])
    print("----------------------------")
    return




def compare_uniform():
    octiles = compute_octiles_uniform()
    print("uniform octiles", octiles)
    O_1, O_2, O_3 = octiles[1], octiles[2], octiles[3]
    O_4, O_5, O_6, O_7 = octiles[4], octiles[5], octiles[6], octiles[7]

    mu, h = 1/2, 1/4
    print("---------------------------")
    print("uniform h:                 ", h)
    print("uniform mu:                ", mu)
    print("uniform median:            ", one_median(octiles))    
    print("uniform 2-quartiles:       ", two_quartiles(octiles))    
    print("uniform 4-quartiles        ", four_quartiles(octiles))    
    print("uniform 4-octiles :        ",four_octiles(octiles))   
    print("uniform all-octiles :      ", all_octiles(octiles))
    print("uniform trap mu:           ", trapezoid(octiles))
    print("uniform 2-rec-2-trap       ", two_rect_two_trap(octiles))
    print("uniform 2-rec-1-trap       ", two_rect_one_trap(octiles))    
    print("uniform simpson:           ", simpson(octiles))
    print("uniform simpson_3:         ", simpson_3(octiles))
    print("uniform simpson_composite: ", simpson_composite(octiles))
    print("----------------------------")
    return
    
    
compare_laplace(0, 1/math.sqrt(2))  # variance-one Laplace; true h = 1/sqrt(2) (~0.707)
# compare_laplace(0, 1)             # optional: unit scale b=1, true h=1
compare_exponential(1)              # rate lambda=1
compare_uniform()
compare_pareto(2.0, 1/2)

```

```python
import math

# Laplace-only C-trapezoid check

def c_trapezoid_method(octiles):
    """C-trapezoid (cubic endpoint + trapezoid) approximation."""
    O_1, O_2, O_3 = octiles[1], octiles[2], octiles[3]
    O_5, O_6, O_7 = octiles[5], octiles[6], octiles[7]

    I1 = (3 * O_1 - 2 * O_2 + 3 * O_3) / 8.0
    I2 = (3 * O_5 - 2 * O_6 + 3 * O_7) / 8.0

    mu = I1 + I2
    h = I2 - I1
    return mu, h

def one_median(octiles):
    return octiles[4], 0

def compute_octiles_laplace(my_mu=0, my_lambda=1):
    O_1 = my_mu + my_lambda * math.log(2 * 1 / 8)
    O_2 = my_mu + my_lambda * math.log(2 * 2 / 8)
    O_3 = my_mu + my_lambda * math.log(2 * 3 / 8)
    O_4 = my_mu + my_lambda * math.log(2 * 4 / 8)
    O_5 = my_mu - my_lambda * math.log(2 - 2 * 5 / 8)
    O_6 = my_mu - my_lambda * math.log(2 - 2 * 6 / 8)
    O_7 = my_mu - my_lambda * math.log(2 - 2 * 7 / 8)
    return [0, O_1, O_2, O_3, O_4, O_5, O_6, O_7]

def compare_laplace(my_mu=0, my_lambda=1):
    octiles = compute_octiles_laplace(my_mu, my_lambda)
    print("laplace octiles", octiles)

    # For Laplace(mu, b), MAD around the median is H=b.
    h = my_lambda
    mu = my_mu

    print("---------------------------")
    print("laplace h:                 ", h)
    print("laplace mu:                ", mu)
    print("laplace median:            ", one_median(octiles))

    mu_c_trap, h_c_trap = c_trapezoid_method(octiles)
    print("laplace c-trapezoid h:     ", h_c_trap)
    print("laplace c-trapezoid mu:    ", mu_c_trap)

    if abs(h) > 1e-10:
        h_error = abs((h - h_c_trap) / h) * 100
        print("laplace h error%:          ", h_error)
    if abs(mu) > 1e-10:
        mu_error = abs((mu - mu_c_trap) / mu) * 100
        print("laplace mu error%:         ", mu_error)
    else:
        print("laplace mu error%:         0")
    print("----------------------------")


# Variance-one Laplace: b = 1/sqrt(2)
compare_laplace(0, 1 / math.sqrt(2))
```
